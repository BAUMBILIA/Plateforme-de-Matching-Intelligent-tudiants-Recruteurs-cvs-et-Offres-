```python
from flask import Flask, render_template_string, request, redirect, url_for, session, jsonify, send_file
from flask import Flask, request
from werkzeug.middleware.proxy_fix import ProxyFix
from datetime import datetime, timedelta
from pymongo import MongoClient
from werkzeug.security import generate_password_hash, check_password_hash
from bson.objectid import ObjectId
from reportlab.lib.pagesizes import letter
from reportlab.lib import colors
from reportlab.platypus import SimpleDocTemplate, Table, TableStyle, Paragraph, Image, Spacer
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.graphics.shapes import Drawing
from reportlab.graphics.charts.piecharts import Pie
from reportlab.graphics.charts.barcharts import VerticalBarChart
import matplotlib.pyplot as plt
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Table, TableStyle, Image
from reportlab.lib.pagesizes import A4
import csv
import requests
import io
import os
import json
from bs4 import BeautifulSoup
import re

app = Flask(__name__)
app.secret_key = 'your_secret_key'

# MongoDB connection setup
client = MongoClient('mongodb://127.0.0.1:27017/recruitment_db')
db = client.recruitment_db
users_collection = db.users
offers_collection = db.offers
scraped_offers_collection = db.scraped_offers
messages_collection = db.messages
cvs_collection = db.cvs
applications_collection = db.applications
projects_collection = db.projects
stats_collection = db.stats
matching_results_collection = db.matching_results
student_profiles_collection = db.student_profiles
notifications_collection = db.notifications

def add_notification(role, message, target_id=None, student_email=None, cv_url=None):
    notifications_collection.insert_one({
        "target_role": role,
        "target_id": target_id,
        "message": message,
        "student_email": student_email,
        "cv_url": cv_url,
        "status": "En cours",
        "timestamp": datetime.utcnow(),
        "read": False
    })


# Templates
HOMEPAGE_TEMPLATE = """
<!DOCTYPE html>
<html lang="fr">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Web4.Jobs - Plateforme de Matching Intelligente</title>
    <style>
        :root {
            --primary-color: #8052e6;
            --primary-dark: #6a40d0;
            --secondary-color: #2c2c6c;
            --accent-color: #ff6b6b;
            --success-color: #4ecdc4;
            --warning-color: #ffe66d;
            --background-color: #f8fafc;
            --card-background: #ffffff;
            --text-primary: #2d3748;
            --text-secondary: #718096;
            --border-color: #e2e8f0;
            --shadow-light: 0 4px 6px rgba(0, 0, 0, 0.07);
            --shadow-medium: 0 10px 25px rgba(0, 0, 0, 0.1);
            --shadow-heavy: 0 20px 40px rgba(0, 0, 0, 0.15);
            --border-radius: 12px;
            --transition: all 0.3s cubic-bezier(0.4, 0, 0.2, 1);
        }
        * {
            margin: 0;
            padding: 0;
            box-sizing: border-box;
        }
        body {
            font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
            background: var(--background-color);
            color: var(--text-primary);
            line-height: 1.6;
            overflow-x: hidden;
        }
        .navbar {
            background: rgba(255, 255, 255, 0.95);
            backdrop-filter: blur(10px);
            padding: 1rem 2rem;
            box-shadow: var(--shadow-light);
            position: fixed;
            top: 0;
            left: 0;
            right: 0;
            z-index: 1000;
            display: flex;
            justify-content: space-between;
            align-items: center;
            animation: fadeInDown 0.8s ease-out;
        }
        .logo {
            height: 50px;
            transition: var(--transition);
            animation: float 3s ease-in-out infinite;
        }
        .logo:hover {
            transform: scale(1.1);
        }
        .hero {
            text-align: center;
            padding: 8rem 2rem 4rem;
            background: linear-gradient(135deg, var(--primary-color), var(--primary-dark), var(--secondary-color));
            background-size: 300% 300%;
            animation: gradientShift 6s ease infinite;
            color: white;
            position: relative;
            overflow: hidden;
        }
        @keyframes gradientShift {
            0% { background-position: 0% 50%; }
            50% { background-position: 100% 50%; }
            100% { background-position: 0% 50%; }
        }
        .hero::before {
            content: '';
            position: absolute;
            top: 0;
            left: 0;
            right: 0;
            bottom: 0;
            background: url('data:image/svg+xml,<svg xmlns="http://www.w3.org/2000/svg" viewBox="0 0 100 100"><defs><pattern id="grain" width="100" height="100" patternUnits="userSpaceOnUse"><circle cx="25" cy="25" r="1" fill="white" opacity="0.1"/><circle cx="75" cy="75" r="1" fill="white" opacity="0.1"/><circle cx="50" cy="10" r="1" fill="white" opacity="0.05"/></pattern></defs><rect width="100" height="100" fill="url(%23grain)"/></svg>');
            pointer-events: none;
        }
        .hero h1 {
            font-size: 3.5rem;
            margin-bottom: 1.5rem;
            font-weight: 700;
            animation: fadeInUp 1s ease-out 0.3s both;
            background: linear-gradient(45deg, #fff, #f0f8ff);
            -webkit-background-clip: text;
            -webkit-text-fill-color: transparent;
            background-clip: text;
        }
        .hero p {
            font-size: 1.3rem;
            max-width: 600px;
            margin: 0 auto 2rem;
            opacity: 0.9;
            animation: fadeInUp 1s ease-out 0.6s both;
        }
        .cta-buttons {
            display: flex;
            gap: 1.5rem;
            justify-content: center;
            margin-top: 2rem;
            animation: fadeInUp 1s ease-out 0.9s both;
            flex-wrap: wrap;
        }
        .cta-button {
            background: rgba(255, 255, 255, 0.2);
            backdrop-filter: blur(10px);
            color: white;
            padding: 1rem 2rem;
            border-radius: var(--border-radius);
            text-decoration: none;
            transition: var(--transition);
            border: 2px solid rgba(255, 255, 255, 0.3);
            font-weight: 600;
            position: relative;
            overflow: hidden;
        }
        .cta-button::before {
            content: '';
            position: absolute;
            top: 0;
            left: -100%;
            width: 100%;
            height: 100%;
            background: linear-gradient(90deg, transparent, rgba(255, 255, 255, 0.4), transparent);
            transition: left 0.5s;
        }
        .cta-button:hover::before {
            left: 100%;
        }
        .cta-button:hover {
            transform: translateY(-2px);
            box-shadow: 0 8px 25px rgba(0, 0, 0, 0.2);
            background: rgba(255, 255, 255, 0.3);
        }
        .cta-button.primary {
            background: var(--accent-color);
            border-color: var(--accent-color);
        }
        .features {
            display: grid;
            grid-template-columns: repeat(auto-fit, minmax(350px, 1fr));
            gap: 2rem;
            padding: 4rem 2rem;
            max-width: 1200px;
            margin: 0 auto;
        }
        .feature-card {
            background: var(--card-background);
            padding: 2.5rem;
            border-radius: var(--border-radius);
            box-shadow: var(--shadow-light);
            transition: var(--transition);
            position: relative;
            overflow: hidden;
            border: 1px solid var(--border-color);
        }
        .feature-card::before {
            content: '';
            position: absolute;
            top: 0;
            left: 0;
            right: 0;
            height: 4px;
            background: linear-gradient(90deg, var(--primary-color), var(--accent-color));
            transform: scaleX(0);
            transition: transform 0.3s ease;
        }
        .feature-card:hover::before {
            transform: scaleX(1);
        }
        .feature-card:hover {
            transform: translateY(-10px) scale(1.02);
            box-shadow: var(--shadow-heavy);
        }
        .feature-card h2 {
            color: var(--primary-color);
            margin-bottom: 1.5rem;
            font-size: 1.5rem;
            display: flex;
            align-items: center;
            gap: 0.5rem;
        }
        .feature-card h2::before {
            content: '✨';
            font-size: 1.2em;
        }
        .feature-card ul {
            list-style: none;
        }
        .feature-card li {
            margin: 1rem 0;
            display: flex;
            align-items: center;
            gap: 0.75rem;
            padding: 0.5rem;
            border-radius: 8px;
            transition: var(--transition);
        }
        .feature-card li:hover {
            background: rgba(128, 82, 230, 0.05);
            transform: translateX(5px);
        }
        .feature-card li::before {
            font-size: 1.2em;
            width: 24px;
            text-align: center;
        }
        .matching-section {
            text-align: center;
            padding: 4rem 2rem;
            background: var(--card-background);
            position: relative;
            overflow: hidden;
        }
        .matching-section::before {
            content: '';
            position: absolute;
            top: -50%;
            left: -50%;
            width: 200%;
            height: 200%;
            background: radial-gradient(circle, rgba(128, 82, 230, 0.05) 0%, transparent 70%);
            animation: rotate 20s linear infinite;
        }
        @keyframes rotate {
            from { transform: rotate(0deg); }
            to { transform: rotate(360deg); }
        }
        .matching-section h2 {
            color: var(--primary-color);
            margin-bottom: 2rem;
            font-size: 2.5rem;
            position: relative;
            z-index: 1;
        }
        .matching-section p {
            font-size: 1.1rem;
            max-width: 600px;
            margin: 0 auto;
            color: var(--text-secondary);
            position: relative;
            z-index: 1;
        }
        @keyframes pulse {
            0%, 100% {
                transform: scale(1);
            }
            50% {
                transform: scale(1.05);
            }
        }
        @keyframes fadeInUp {
            from {
                opacity: 0;
                transform: translateY(30px);
            }
            to {
                opacity: 1;
                transform: translateY(0);
            }
        }
        @keyframes slideInLeft {
            from {
                opacity: 0;
                transform: translateX(-50px);
            }
            to {
                opacity: 1;
                transform: translateX(0);
            }
        }
        @keyframes slideInRight {
            from {
                opacity: 0;
                transform: translateX(50px);
            }
            to {
                opacity: 1;
                transform: translateX(0);
            }
        }
        @keyframes fadeInDown {
            from {
                opacity: 0;
                transform: translateY(-30px);
            }
            to {
                opacity: 1;
                transform: translateY(0);
            }
        }
    </style>
</head>
<body>
    <nav class="navbar">
        <img src="https://w4j.yool.education/assets/web4jobs/images/logo.png" alt="Web4.Jobs Logo" class="logo">
        <button id="loginBtn" class="cta-button primary" onclick="window.location.href='/login'">Se connecter</button>
    </nav>

    <div id="homepage" class="page">
        <section class="hero">
            <h1>Trouvez le match parfait</h1>
            <p>Plateforme intelligente de matching entre étudiants talentueux et opportunités de freelance</p>
            <div class="cta-buttons">
                <button class="cta-button" data-scroll-to="students">Je suis étudiant</button>
                <button class="cta-button" data-scroll-to="recruiters">Je suis recruteur</button>
            </div>
        </section>

        <section class="features">
            <div class="feature-card animate-on-scroll" id="students">
                <h2>Pour les Étudiants</h2>
                <ul>
                    <li>📄 Uploadez votre CV en quelques clics</li>
                    <li>🎨 Créez un portfolio professionnel</li>
                    <li>💼 Découvrez des opportunités personnalisées</li>
                    <li>🚀 Développez votre carrière freelance</li>
                    <li>📈 Suivez vos candidatures en temps réel</li>
                </ul>
            </div>

            <div class="feature-card animate-on-scroll" id="recruiters">
                <h2>Pour les Recruteurs</h2>
                <ul>
                    <li>📢 Publiez vos offres facilement</li>
                    <li>🔍 Découvrez des talents prometteurs</li>
                    <li>📊 Gérez toutes vos candidatures</li>
                    <li>⚡ Recrutement rapide et efficace</li>
                    <li>🎯 Matching basé sur les compétences</li>
                </ul>
            </div>
        </section>

        <section class="matching-section animate-on-scroll">
            <h2>Matching Intelligent Alimenté par l'IA</h2>
            <p>Notre algorithme avancé analyse les compétences, l'expérience et les préférences pour recommander les meilleures correspondances entre talents et opportunités.</p>
        </section>
    </div>
</body>
</html>
"""

LOGIN_TEMPLATE = """
<!DOCTYPE html>
<html lang="fr">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Connexion - Web4.Jobs</title>
    <style>
        :root {
            --primary-color: #8052e6;
            --primary-dark: #6a40d0;
            --secondary-color: #2c2c6c;
            --accent-color: #ff6b6b;
            --success-color: #4ecdc4;
            --warning-color: #ffe66d;
            --background-color: #f8fafc;
            --card-background: #ffffff;
            --text-primary: #2d3748;
            --text-secondary: #718096;
            --border-color: #e2e8f0;
            --shadow-light: 0 4px 6px rgba(0, 0, 0, 0.07);
            --shadow-medium: 0 10px 25px rgba(0, 0, 0, 0.1);
            --shadow-heavy: 0 20px 40px rgba(0, 0, 0, 0.15);
            --border-radius: 12px;
            --transition: all 0.3s cubic-bezier(0.4, 0, 0.2, 1);
        }
        html, body {
            margin: 0;
            padding: 0;
            height: 100%;
        }
        * {
            box-sizing: border-box;
        }
        body {
            font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
            background: transparent;
            display: flex;
            flex-direction: column;
            justify-content: center;
            align-items: center;
            min-height: 100vh;
            padding: 2rem;
            position: relative;
            z-index: 1;
        }
        #lottie-bg {
            position: fixed;
            top: 0;
            left: 0;
            z-index: -1;
            width: 100vw;
            height: 100vh;
            opacity: 0.15;
            pointer-events: none;
        }
        .overlay {
            background: rgba(255, 255, 255, 0.8);
            position: relative;
            z-index: 1;
            min-height: 100vh;
            padding: 20px;
        }
        .logo-container {
            margin-bottom: 2rem;
            animation: pulse 2s ease-in-out infinite;
        }
        .logo {
            height: 80px;
            transition: var(--transition);
        }
        .logo:hover {
            transform: scale(1.1);
        }
        .login-container {
            background: rgba(255, 255, 255, 0.8);
            backdrop-filter: blur(5px);
            border-radius: 8px;
            box-shadow: var(--shadow-heavy);
            width: 100%;
            max-width: 360px;
            margin: 0 auto;
            overflow: hidden;
            border: 1px solid rgba(255, 255, 255, 0.2);
            animation: fadeInUp 0.8s ease-out;
            position: relative;
            padding: 0; /* On retire le padding ici car il sera géré par .form-content */
        }
        .login-header {
            text-align: center;
            background: linear-gradient(135deg, var(--primary-color), var(--primary-dark));
            color: white;
            padding: 2rem;
        }
        .login-header h1 {
            margin: 0;
            font-size: 1.8rem;
            font-weight: 600;
        }
        .login-header p {
            margin: 0.5rem 0 0;
            opacity: 0.9;
            font-size: 0.9rem;
        }
        .form-content {
            padding: 2rem;
            width: 100%;
        }
        .login-form {
            position: relative;
            z-index: 1;
        }
        .form-group {
            margin-bottom: 1.5rem;
            position: relative;
        }
        .form-group label {
            display: block;
            margin-bottom: 0.5rem;
            font-weight: 600;
            color: var(--text-primary);
            font-size: 0.9rem;
        }
        .form-group select,
        .form-group input {
            width: 100%;
            padding: 1rem;
            border: 2px solid var(--border-color);
            border-radius: var(--border-radius);
            font-size: 1rem;
            transition: var(--transition);
            background: white;
        }
        .form-group select:focus,
        .form-group input:focus {
            outline: none;
            border-color: var(--primary-color);
            box-shadow: 0 0 0 3px rgba(128, 82, 230, 0.1);
        }
        .login-button {
            width: 100%;
            padding: 1rem;
            background: linear-gradient(135deg, var(--primary-color), var(--primary-dark));
            color: white;
            border: none;
            border-radius: var(--border-radius);
            font-size: 1.1rem;
            font-weight: 600;
            cursor: pointer;
            transition: var(--transition);
            position: relative;
            overflow: hidden;
        }
        .login-button::before {
            content: '';
            position: absolute;
            top: 0;
            left: -100%;
            width: 100%;
            height: 100%;
            background: linear-gradient(90deg, transparent, rgba(255, 255, 255, 0.2), transparent);
            transition: left 0.5s;
        }
        .login-button:hover::before {
            left: 100%;
        }
        .login-button:hover {
            transform: translateY(-2px);
            box-shadow: var(--shadow-medium);
        }
        .login-footer {
            margin-top: 2rem;
            text-align: center;
            color: white;
            opacity: 0.8;
            font-size: 0.9rem;
            position: relative;
            z-index: 1;
        }
        .register-link {
            text-align: center;
            margin-top: 1rem;
        }
        .register-link a {
            color: var(--primary-color);
            text-decoration: none;
            font-weight: 600;
        }
        .register-link a:hover {
            text-decoration: underline;
        }
        @keyframes pulse {
            0%, 100% {
                transform: scale(1);
            }
            50% {
                transform: scale(1.05);
            }
        }
        @keyframes fadeInUp {
            from {
                opacity: 0;
                transform: translateY(30px);
            }
            to {
                opacity: 1;
                transform: translateY(0);
            }
        }
    </style>
</head>
<body>
    <div id="lottie-bg"></div>

    <div class="logo-container">
        <img src="https://w4j.yool.education/assets/web4jobs/images/logo.png" alt="Web4.Jobs Logo" class="logo">
    </div>

    <div class="login-container">
        <div class="login-header">
            <h1>Bienvenue</h1>
            <p>Connectez-vous à la plateforme Web4.Jobs</p>
        </div>

        <div class="form-content">
            <form id="loginForm" class="login-form" action="/authenticate" method="POST">
                <div class="form-group">
                    <label for="email">E-mail</label>
                    <input type="email" id="email" name="email" placeholder="Entrez votre e-mail" required>
                </div>

                <div class="form-group">
                    <label for="password">Mot de passe</label>
                    <input type="password" id="password" name="password" placeholder="Entrez votre mot de passe" required>
                </div>

                <button type="submit" class="login-button">
                    <span>Se connecter</span>
                </button>
            </form>

            <div class="register-link">
                <p>Vous n'avez pas de compte ? <a href="/register">Inscription</a></p>
            </div>
        </div>
    </div>

    <div class="login-footer">
        <p>© 2025 Web4.Jobs. Tous droits réservés.</p>
    </div>

    <!-- Intégration de Lottie -->
    <script src="https://unpkg.com/lottie-web@5.7.4/build/player/lottie.min.js"></script>
    <script>
        document.addEventListener('DOMContentLoaded', function() {
            lottie.loadAnimation({
                container: document.getElementById('lottie-bg'),
                renderer: 'svg',
                loop: true,
                autoplay: true,
                path: 'https://assets4.lottiefiles.com/packages/lf20_dygyj2yd.json'
            });
        });
    </script>
</body>
</html>

"""

REGISTER_TEMPLATE = """
<!DOCTYPE html>
<html lang="fr">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Inscription - Web4.Jobs</title>
    <style>
        :root {
            --primary-color: #2c2c6c;
            --primary-dark: #1a1a40;
            --secondary-color: #8052e6;
            --accent-color: #ff6b6b;
            --success-color: #4ecdc4;
            --warning-color: #ffe66d;
            --background-color: #f8fafc;
            --card-background: #ffffff;
            --text-primary: #2d3748;
            --text-secondary: #718096;
            --border-color: #e2e8f0;
            --shadow-light: 0 4px 6px rgba(0, 0, 0, 0.07);
            --shadow-medium: 0 10px 25px rgba(0, 0, 0, 0.1);
            --shadow-heavy: 0 20px 40px rgba(0, 0, 0, 0.15);
            --border-radius: 12px;
            --transition: all 0.3s cubic-bezier(0.4, 0, 0.2, 1);
        }
        * {
            margin: 0;
            padding: 0;
            box-sizing: border-box;
        }
        body {
            font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
            background-color: #ffffff;
            display: flex;
            flex-direction: column;
            justify-content: center;
            align-items: center;
            min-height: 100vh;
            padding: 2rem;
            position: relative;
            overflow-x: hidden;
        }

        .dot {
            position: absolute;
            width: 20px;
            height: 20px;
            background-color: var(--primary-color);
            border-radius: 50%;
            opacity: 0.8;
            animation: blink 1.5s infinite alternate;
            cursor: pointer;
            transition: opacity 0.3s ease;
        }

        @keyframes blink {
            from {
                opacity: 0.8;
            }
            to {
                opacity: 0.2;
            }
        }

        .logo-container {
            margin-bottom: 2rem;
            animation: pulse 2s ease-in-out infinite;
        }
        .logo {
            height: 80px;
            transition: var(--transition);
        }
        .logo:hover {
            transform: scale(1.1);
        }
        .register-container {
            background: rgba(255, 255, 255, 0.95);
            backdrop-filter: blur(20px);
            border-radius: var(--border-radius);
            box-shadow: var(--shadow-heavy);
            width: 90%;
            max-width: 400px;
            overflow: hidden;
            border: 1px solid rgba(255, 255, 255, 0.2);
            animation: fadeInUp 0.8s ease-out;
        }
        .register-header {
            text-align: center;
            background: linear-gradient(135deg, var(--secondary-color), var(--primary-dark));
            color: white;
            padding: 2rem;
        }
        .register-header h1 {
            margin: 0;
            font-size: 1.8rem;
            font-weight: 600;
        }
        .register-header p {
            margin: 0.5rem 0 0;
            opacity: 0.9;
            font-size: 0.9rem;
        }
        .form-content {
            padding: 2rem;
        }
        .form-group {
            margin-bottom: 1.5rem;
            position: relative;
        }
        .form-group label {
            display: block;
            margin-bottom: 0.5rem;
            font-weight: 600;
            color: var(--text-primary);
            font-size: 0.9rem;
        }
        .form-group select,
        .form-group input {
            width: 100%;
            padding: 1rem;
            border: 2px solid var(--border-color);
            border-radius: var(--border-radius);
            font-size: 1rem;
            transition: var(--transition);
            background: white;
        }
        .form-group select:focus,
        .form-group input:focus {
            outline: none;
            border-color: var(--secondary-color);
            box-shadow: 0 0 0 3px rgba(128, 82, 230, 0.1);
        }
        .register-button {
            width: 100%;
            padding: 1rem;
            background: linear-gradient(135deg, var(--secondary-color), var(--primary-dark));
            color: white;
            border: none;
            border-radius: var(--border-radius);
            font-size: 1.1rem;
            font-weight: 600;
            cursor: pointer;
            transition: var(--transition);
            position: relative;
            overflow: hidden;
        }
        .register-button::before {
            content: '';
            position: absolute;
            top: 0;
            left: -100%;
            width: 100%;
            height: 100%;
            background: linear-gradient(90deg, transparent, rgba(255, 255, 255, 0.2), transparent);
            transition: left 0.5s;
        }
        .register-button:hover::before {
            left: 100%;
        }
        .register-button:hover {
            transform: translateY(-2px);
            box-shadow: var(--shadow-medium);
        }
        .login-footer {
            margin-top: 2rem;
            text-align: center;
            color: var(--text-primary);
            opacity: 0.8;
            font-size: 0.9rem;
        }
        .login-link {
            text-align: center;
            margin-top: 1rem;
        }
        .login-link a {
            color: var(--secondary-color);
            text-decoration: none;
            font-weight: 600;
        }
        .login-link a:hover {
            text-decoration: underline;
        }
        @keyframes pulse {
            0%, 100% {
                transform: scale(1);
            }
            50% {
                transform: scale(1.05);
            }
        }
        @keyframes fadeInUp {
            from {
                opacity: 0;
                transform: translateY(30px);
            }
            to {
                opacity: 1;
                transform: translateY(0);
            }
        }
    </style>
</head>
<body>
    <div class="logo-container">
        <img src="https://w4j.yool.education/assets/web4jobs/images/logo.png" alt="Web4.Jobs Logo" class="logo">
    </div>

    <div class="register-container">
        <div class="register-header">
            <h1>Inscription</h1>
            <p>Créez un compte sur la plateforme Web4.Jobs</p>
        </div>

        <div class="form-content">
            <form id="registerForm" class="register-form" action="/register" method="POST">
                <div class="form-group">
                    <label for="email">E-mail</label>
                    <input type="email" id="email" name="email" placeholder="Entrez votre e-mail" required>
                </div>

                <div class="form-group">
                    <label for="full_name">Nom Complet</label>
                    <input type="text" id="full_name" name="full_name" placeholder="Entrez votre nom complet" required>
                </div>

                <div class="form-group">
                    <label for="user_type">Type d'utilisateur</label>
                    <select id="user_type" name="user_type" required>
                        <option value="" disabled selected>Sélectionnez votre profil</option>
                        <option value="freelance">👨‍💻 Freelance</option>
                        <option value="recruiter">🏢 Recruteur</option>
                        <option value="admin">⚙️ Administrateur</option>
                    </select>
                </div>

                <div class="form-group">
                    <label for="password">Mot de passe</label>
                    <input type="password" id="password" name="password" placeholder="Entrez votre mot de passe" required>
                </div>

                <button type="submit" class="register-button">
                    <span>S'inscrire</span>
                </button>
            </form>

            <div class="login-link">
                <p>Vous avez déjà un compte ? <a href="/login">Connexion</a></p>
            </div>
        </div>
    </div>

    <div class="login-footer">
        <p>© 2025 Web4.Jobs. Tous droits réservés.</p>
    </div>

    <!-- Ajout des taches -->
    <div class="dot" style="top: 10%; left: 20%;"></div>
    <div class="dot" style="top: 30%; left: 50%;"></div>
    <div class="dot" style="top: 60%; left: 80%;"></div>
    <div class="dot" style="top: 20%; left: 70%;"></div>
    <div class="dot" style="top: 80%; left: 30%;"></div>
    <div class="dot" style="top: 40%; left: 60%;"></div>
    <div class="dot" style="top: 50%; left: 10%;"></div>
    <div class="dot" style="top: 70%; left: 40%;"></div>

    <script>
        document.addEventListener('DOMContentLoaded', function() {
            const dots = document.querySelectorAll('.dot');

            dots.forEach(dot => {
                dot.addEventListener('click', function() {
                    this.style.opacity = '0';
                    setTimeout(() => {
                        this.remove();
                    }, 1000);
                });
            });
        });
    </script>
</body>
</html>


"""

STUDENT_DASHBOARD_TEMPLATE = """
<!DOCTYPE html>
<html lang="fr">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Tableau de bord Étudiant</title>
    <style>
        :root {
            --primary-color: #8052e6;
            --primary-dark: #6a40d0;
            --secondary-color: #2c2c6c;
            --accent-color: #ff6b6b;
            --success-color: #4ecdc4;
            --warning-color: #ffe66d;
            --background-color: #f8fafc;
            --card-background: #f5f5f5;
            --text-primary: #2d3748;
            --text-secondary: #718096;
            --border-color: #e2e8f0;
            --shadow-light: 0 4px 6px rgba(0, 0, 0, 0.07);
            --shadow-medium: 0 10px 25px rgba(0, 0, 0, 0.1);
            --shadow-heavy: 0 20px 40px rgba(0, 0, 0, 0.15);
            --border-radius: 12px;
            --transition: all 0.3s cubic-bezier(0.4, 0, 0.2, 1);
        }
        * {
            margin: 0;
            padding: 0;
            box-sizing: border-box;
        }
        body {
            font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
            background: var(--background-color);
            color: var(--text-primary);
            line-height: 1.6;
        }
        .dashboard {
            min-height: 100vh;
            padding-top: 80px;
        }
        .dashboard-header {
            background: linear-gradient(135deg, var(--secondary-color), var(--primary-color));
            color: white;
            padding: 2rem;
            margin: 2rem;
            border-radius: var(--border-radius);
            box-shadow: var(--shadow-medium);
            display: flex;
            justify-content: space-between;
            align-items: center;
            animation: slideInDown 0.6s ease-out;
        }
        .header-actions {
            display: flex;
            gap: 1rem;
        }
        .dashboard-content {
            display: grid;
            grid-template-columns: repeat(auto-fit, minmax(350px, 1fr));
            gap: 2rem;
            padding: 0 2rem 2rem;
        }
        .dashboard-card {
            background: var(--card-background);
            padding: 2rem;
            border-radius: var(--border-radius);
            box-shadow: var(--shadow-light);
            border: 1px solid var(--border-color);
            transition: var(--transition);
            animation: fadeInUp 0.6s ease-out;
        }
        .dashboard-card:hover {
            transform: translateY(-5px);
            box-shadow: var(--shadow-medium);
        }
        .dashboard-card h2 {
            color: var(--primary-color);
            margin-bottom: 1rem;
            display: flex;
            align-items: center;
            gap: 0.5rem;
        }
        .action-button {
            background: linear-gradient(135deg, var(--primary-color), var(--primary-dark));
            color: white;
            border: none;
            padding: 0.75rem 1.5rem;
            border-radius: var(--border-radius);
            font-weight: 600;
            cursor: pointer;
            transition: var(--transition);
            position: relative;
            overflow: hidden;
            margin: 0.5rem 0;
        }
        .action-button:hover {
            transform: translateY(-2px);
            box-shadow: var(--shadow-medium);
        }
        .action-button:active {
            transform: translateY(0);
        }
        .stats-grid {
            display: grid;
            grid-template-columns: repeat(auto-fit, minmax(150px, 1fr));
            gap: 1rem;
            margin: 1rem 0;
        }
        .stat-item {
            text-align: center;
            padding: 1rem;
            background: rgba(128, 82, 230, 0.05);
            border-radius: var(--border-radius);
            transition: var(--transition);
        }
        .stat-item:hover {
            background: rgba(128, 82, 230, 0.1);
            transform: scale(1.05);
        }
        .stat-number {
            font-size: 2rem;
            font-weight: 700;
            color: var(--primary-color);
            display: block;
        }
        .stat-label {
            font-size: 0.9rem;
            color: var(--text-secondary);
            margin-top: 0.5rem;
        }
        .upload-area {
            border: 2px dashed var(--border-color);
            border-radius: var(--border-radius);
            padding: 2rem;
            text-align: center;
            transition: var(--transition);
            cursor: pointer;
            margin: 1rem 0;
        }
        .upload-area:hover {
            border-color: var(--primary-color);
            background: rgba(128, 82, 230, 0.05);
        }
        .upload-area.dragover {
            border-color: var(--success-color);
            background: rgba(78, 205, 196, 0.1);
        }
        .file-preview {
            margin-top: 1rem;
            padding: 1rem;
            border: 1px solid var(--border-color);
            border-radius: var(--border-radius);
            background-color: rgba(128, 82, 230, 0.05);
        }
        .ai-assistant-container {
            display: flex;
            flex-direction: column;
            gap: 10px;
            margin-bottom: 1rem;
        }
        .ai-input {
            flex: 1;
            padding: 12px 20px;
            border: 2px solid var(--border-color);
            border-radius: var(--border-radius);
            font-size: 1rem;
            transition: var(--transition);
            outline: none;
            width: 100%;
        }
        .ai-input:focus {
            border-color: var(--primary-color);
            box-shadow: 0 0 0 3px rgba(128, 82, 230, 0.2);
        }
        .button-container {
            display: flex;
            gap: 10px;
            flex-wrap: wrap;
        }
        .ai-button {
            background: linear-gradient(135deg, var(--primary-color), var(--primary-dark));
            color: white;
            border: none;
            padding: 12px 24px;
            border-radius: var(--border-radius);
            font-weight: 600;
            cursor: pointer;
            transition: var(--transition);
            white-space: nowrap;
            font-size: 1rem;
            flex: 1;
        }
        .ai-button:hover {
            background: linear-gradient(135deg, var(--primary-dark), var(--primary-color));
            transform: translateY(-2px);
            box-shadow: var(--shadow-medium);
        }
        .ai-button.secondary {
            background: linear-gradient(135deg, #4facfe 0%, #00f2fe 100%);
            box-shadow: 0 8px 25px rgba(79, 172, 254, 0.3);
        }
        .ai-button.secondary:hover {
            box-shadow: 0 12px 35px rgba(79, 172, 254, 0.4);
        }
        .ai-response {
            margin-top: 1rem;
            padding: 1rem;
            border-radius: var(--border-radius);
            background-color: rgba(128, 82, 230, 0.05);
            border: 1px solid var(--border-color);
            min-height: 50px;
        }
        .project-item {
            background: rgba(245, 245, 245, 0.9);
            border-radius: var(--border-radius);
            padding: 1rem;
            margin-bottom: 1rem;
            border: 1px solid var(--border-color);
            box-shadow: var(--shadow-light);
        }
        @media (max-width: 768px) {
            .dashboard-content {
                grid-template-columns: 1fr;
                padding: 0 1rem 1rem;
            }
            .dashboard-header {
                flex-direction: column;
                text-align: center;
                gap: 1rem;
            }
            .header-actions {
                flex-direction: column;
                gap: 0.5rem;
                width: 100%;
            }
            .header-actions .action-button {
                width: 100%;
            }
        }
        @keyframes fadeInUp {
            from {
                opacity: 0;
                transform: translateY(30px);
            }
            to {
                opacity: 1;
                transform: translateY(0);
            }
        }
        @keyframes slideInDown {
            from {
                opacity:  0;
                transform: translateY(-30px);
            }
            to {
                opacity: 1;
                transform: translateY(0);
            }
        }
    </style>
</head>
<body>
    <div class="dashboard">
        <div class="dashboard-header">
            <h1>🎓 Tableau de bord Étudiant</h1>
            <div class="header-actions">
                <button id="btnAddProject" class="action-button">Ajouter un projet</button>
                <button id="postulerBtn" class="action-button" onclick="window.location.href='/postuler'">Postuler</button>
                <button id="messagesBtn" class="action-button" onclick="window.location.href='/messages'">Messagerie</button>
                <button id="logoutBtn" class="action-button" onclick="window.location.href='/logout'">Déconnexion</button>
            </div>
        </div>

        <div class="dashboard-content">
            <div class="dashboard-card">
                <h2>📄 Mon CV</h2>
                <p>Vous n'avez pas encore téléversé de CV. Ajoutez-en un pour pouvoir postuler aux offres.</p>

                <div class="upload-area" id="cvUploadArea">
                    <input type="file" id="cvFileInput" accept=".pdf,.doc,.docx" style="opacity: 0; width: 100%; height: 100%; position: absolute; top: 0; left: 0; cursor: pointer;">
                    <div style="z-index: 1; position: relative;">
                        <p>📤 <strong>Upload un fichier</strong><br>
                        Glissez-déposez ici ou cliquez pour sélectionner un fichier (.pdf, .docx, .doc)</p>
                    </div>
                </div>
                <div id="filePreview" class="file-preview" style="display: none;">
                    <p>Fichier sélectionné: <span id="fileName"></span></p>
                </div>
                <div id="cvFileStatus" style="margin-top: 1rem;"></div>

                <div class="stat-item">
                    <span class="stat-number" id="totalApplications">{{ total_applications }}</span>
                    <span class="stat-label">Candidatures</span>
                </div>
            </div>

            <div class="dashboard-card">
                <h2>💼 Offres Recommandées</h2>
                <p>Téléchargez votre CV pour voir des offres personnalisées basées sur vos compétences.</p>
                <button class="action-button" id="fetchOffersBtn">📦 Obtenir des Offres</button>
                <div id="recommendedOffers"></div>
            </div>

            <div class="dashboard-card">
                <h2>💬 Assistant IA</h2>
                <div class="ai-assistant-container">
                    <input type="text" id="chatPrompt" class="ai-input" placeholder="Posez une question sur votre profil..." />

                    <div class="button-container">
                        <button class="ai-button" onclick="askAI()">
                            💬 Interroger
                        </button>
                        <button class="ai-button secondary" onclick="askAIWithMatching()">
                            🎯 Interroger avec matching
                        </button>
                    </div>

                    <div id="chatResponse" class="ai-response"></div>
                </div>
            </div>

            <div class="dashboard-card">
                <button class="action-button" id="btnMatching">🎯 Lancer le matching avec les offres</button>
                <div id="matchingResults" style="margin-top: 1rem; padding: 1rem; background: #f8f9fa; border-radius: 10px; display: none;">
                    <h3>Résultats du Matching</h3>
                    <div id="matchingDetails"></div>
                </div>
            </div>
        </div>
    </div>
    <script>
        document.addEventListener("DOMContentLoaded", function () {
            const uploadArea = document.getElementById("cvUploadArea");
            const fileInput = document.getElementById("cvFileInput");
            const filePreview = document.getElementById("filePreview");
            const fileNameSpan = document.getElementById("fileName");

            uploadArea.addEventListener("click", () => {
                fileInput.click();
            });

            uploadArea.addEventListener("dragover", (event) => {
                event.preventDefault();
                uploadArea.classList.add("dragover");
            });

            uploadArea.addEventListener("dragleave", () => {
                uploadArea.classList.remove("dragover");
            });

            uploadArea.addEventListener("drop", (event) => {
                event.preventDefault();
                uploadArea.classList.remove("dragover");
                const files = event.dataTransfer.files;
                if (files.length) {
                    fileInput.files = files;
                    updateFilePreview(files[0]);
                }
            });

            fileInput.addEventListener("change", function () {
                const files = this.files;
                if (files.length) {
                    updateFilePreview(files[0]);
                    const formData = new FormData();
                    formData.append("cv", files[0]);

                    fetch("/upload_cv", {
                        method: "POST",
                        body: formData
                    })
                    .then(res => res.json())
                    .then(data => {
                        if (data.success) {
                            document.getElementById("cvFileStatus").innerHTML = `
                                ✅ Dernier CV : <strong>${data.filename}</strong>
                            `;
                        } else {
                            alert("❌ Erreur : " + data.message);
                        }
                    });
                }
            });

            function updateFilePreview(file) {
                fileNameSpan.textContent = file.name;
                filePreview.style.display = "block";
            }

            document.getElementById("btnAddProject").addEventListener("click", function () {
                const title = prompt("Titre du projet :");
                const description = prompt("Description :");
                const technologies = prompt("Technos (séparées par virgule) :");
                const url = prompt("Lien vers le projet (GitHub, etc.) :");

                if (title && description && technologies && url) {
                    const formData = new FormData();
                    formData.append("title", title);
                    formData.append("description", description);
                    formData.append("technologies", technologies);
                    formData.append("url", url);

                    fetch("/add_project", {
                        method: "POST",
                        body: formData
                    })
                    .then(res => res.json())
                    .then(data => {
                        alert("🚀 Projet ajouté !");
                    });
                } else {
                    alert("Veuillez remplir tous les champs pour ajouter un projet.");
                }
            });

            document.getElementById("btnMatching").addEventListener("click", function () {
                fetch("/match_cv_to_offers", { method: "POST" })
                .then(res => res.json())
                .then(data => {
                    if (data.error) {
                        alert('Erreur: ' + data.error);
                        return;
                    }

                    const result = data.match_result;
                    const offres = data.offers_used || [];
                    let html = `
                        <p><strong>🔍 Matching basé sur :</strong> ${offres.join(", ")}</p>
                        <div style="white-space:pre-wrap; margin-top:1rem;">${result}</div>
                    `;

                    document.getElementById('matchingDetails').innerHTML = html;
                    document.getElementById('matchingResults').style.display = 'block';
                })
                .catch(error => {
                    console.error('Error:', error);
                    alert('Une erreur est survenue lors du matching');
                });
            });

            document.getElementById("fetchOffersBtn").addEventListener("click", () => {
                fetch('/api/get_recent_offers')
                .then(res => res.json())
                .then(data => {
                    const container = document.getElementById('recommendedOffers');
                    if (!data.length) {
                        container.innerHTML = "<p>😕 Aucune offre disponible pour l’instant.</p>";
                        return;
                    }
                    container.innerHTML = data.map(o => `
                        <div style="border:1px solid #ddd; border-radius:10px; padding:1rem; margin-top:1rem;">
                            <strong>${o.title}</strong><br>
                            <em>${o.description}</em><br>
                            <small>Tags : ${o.tags.join(", ")}</small>
                        </div>
                    `).join('');
                });
            });

            fetch("/get_latest_cv")
            .then(res => res.json())
            .then(data => {
                if (data && data.filename) {
                    const cvArea = document.getElementById("cvUploadArea");
                    cvArea.innerHTML = `<p>📎 Dernier CV : <strong>${data.filename}</strong></p>`;
                }
            });
        });

        function loadRecommendedOffers(skill = "python") {
            fetch('/matching_offers/' + skill)
            .then(res => res.json())
            .then(data => {
                const container = document.getElementById('recommendedOffers');
                if (!container) return;

                if (data.matches.length === 0) {
                    container.innerHTML = '<p>🙁 Aucune offre ne correspond pour l’instant.</p>';
                    return;
                }

                container.innerHTML = data.matches.map(o => `
                    <div style="border:1px solid #ccc; border-radius:10px; padding:1rem; margin-bottom:1rem;">
                        <h3>${o.title}</h3>
                        <p>${o.description}</p>
                        <small>Tags : ${o.matching_tags.join(', ')}</small>
                    </div>
                `).join('');
            });
        }

        document.addEventListener("DOMContentLoaded", function () {
            loadRecommendedOffers("python");
        });

        function askAI() {
            const msg = document.getElementById("chatPrompt").value;
            const responseDiv = document.getElementById("chatResponse");

            if (!msg.trim()) {
                responseDiv.innerText = "Veuillez saisir une question.";
                return;
            }

            responseDiv.className = "ai-response loading";
            responseDiv.innerText = "";

            fetch("/ask_ai", {
                method: "POST",
                headers: { "Content-Type": "application/json" },
                body: JSON.stringify({ question: msg })
            })
            .then(res => res.json())
            .then(data => {
                responseDiv.className = "ai-response";
                responseDiv.innerText = data.answer;
                document.getElementById("chatPrompt").value = "";
            })
            .catch(error => {
                responseDiv.className = "ai-response";
                responseDiv.innerText = "Erreur lors de la communication avec l'assistant.";
            });
        }

        function askAIWithMatching() {
            const msg = document.getElementById("chatPrompt").value;
            const responseDiv = document.getElementById("chatResponse");

            if (!msg.trim()) {
                responseDiv.innerText = "Veuillez saisir une question.";
                return;
            }

            responseDiv.className = "ai-response loading";
            responseDiv.innerText = "";

            fetch("/ask_ai_with_matching", {
                method: "POST",
                headers: { "Content-Type": "application/json" },
                body: JSON.stringify({ question: msg })
            })
            .then(res => res.json())
            .then(data => {
                responseDiv.className = "ai-response";
                responseDiv.innerText = data.answer;
                document.getElementById("chatPrompt").value = "";
            })
            .catch(error => {
                responseDiv.className = "ai-response";
                responseDiv.innerText = "Erreur lors de la communication avec l'assistant.";
            });
        }

        document.getElementById("chatPrompt").addEventListener("keypress", function(event) {
            if (event.key === "Enter") {
                askAI();
            }
        });
    </script>
</body>
</html>



"""

RECRUITER_DASHBOARD_TEMPLATE = """
<!DOCTYPE html>
<html lang="fr">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Tableau de bord Recruteur</title>
    <style>
        :root {
            --primary-color: #667eea;
            --secondary-color: #764ba2;
            --accent-color: #f093fb;
            --success-color: #4facfe;
            --warning-color: #43e97b;
            --danger-color: #fa709a;
            --dark-color: #2c3e50;
            --light-color: #ecf0f1;
            --text-color: #34495e;
            --card-shadow: 0 10px 30px rgba(0, 0, 0, 0.1);
            --gradient-bg: linear-gradient(135deg, var(--primary-color), var(--secondary-color));
            --glass-bg: rgba(255, 255, 255, 0.1);
            --glass-border: rgba(255, 255, 255, 0.2);
        }
        * {
            margin: 0;
            padding: 0;
            box-sizing: border-box;
        }
        body {
            font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
            background: var(--gradient-bg);
            min-height: 100vh;
            color: var(--text-color);
            line-height: 1.6;
        }
        .dashboard {
            min-height: 100vh;
            display: flex;
            flex-direction: column;
        }
        .dashboard-header {
            background: var(--glass-bg);
            backdrop-filter: blur(20px);
            border-bottom: 1px solid var(--glass-border);
            padding: 1.5rem 2rem;
            display: flex;
            justify-content: space-between;
            align-items: center;
            box-shadow: var(--card-shadow);
            position: sticky;
            top: 0;
            z-index: 100;
        }
        .dashboard-header h1 {
            color: white;
            font-size: 1.8rem;
            font-weight: 600;
            text-shadow: 0 2px 10px rgba(0, 0, 0, 0.3);
        }
        .cta-button {
            background: linear-gradient(45deg, var(--accent-color), var(--danger-color));
            color: white;
            border: none;
            padding: 0.8rem 1.5rem;
            border-radius: 25px;
            font-weight: 600;
            cursor: pointer;
            transition: all 0.3s ease;
            box-shadow: 0 5px 15px rgba(0, 0, 0, 0.2);
        }
        .cta-button:hover {
            transform: translateY(-2px);
            box-shadow: 0 8px 25px rgba(0, 0, 0, 0.3);
        }
        .dashboard-content {
            flex: 1;
            padding: 2rem;
            display: grid;
            grid-template-columns: 1fr 300px;
            gap: 2rem;
            max-width: 1400px;
            margin: 0 auto;
            width: 100%;
            background: rgba(255, 255, 255, 0.9);
            backdrop-filter: blur(8px);
            border-radius: 15px;
            box-shadow: 0 10px 30px rgba(0, 0, 0, 0.1);
            overflow: hidden;
        }
        .main-content {
            display: flex;
            flex-direction: column;
            gap: 2rem;
        }
        .sidebar {
            display: flex;
            flex-direction: column;
            gap: 1.5rem;
            background-color: white;
            padding: 1rem;
            border-radius: 10px;
        }
        .dashboard-card {
            background: rgba(255, 255, 255, 0.95);
            backdrop-filter: blur(10px);
            border-radius: 15px;
            padding: 1.5rem;
            box-shadow: 0 8px 20px rgba(0, 0, 0, 0.08);
            border: 1px solid rgba(255, 255, 255, 0.3);
            transition: transform 0.3s ease, box-shadow 0.3s ease;
            margin-bottom: 1.5rem;
            border-top: 3px solid var(--primary-color);
        }
        .dashboard-card:hover {
            transform: translateY(-5px);
            box-shadow: 0 12px 30px rgba(0, 0, 0, 0.12);
        }
        .dashboard-card h2 {
            color: var(--dark-color);
            margin-bottom: 1.5rem;
            font-size: 1.4rem;
            font-weight: 600;
        }
        .stats-grid {
            display: grid;
            grid-template-columns: repeat(auto-fit, minmax(150px, 1fr));
            gap: 1.5rem;
            margin-bottom: 2rem;
        }
        .stat-item {
            text-align: center;
            padding: 1.5rem;
            background: linear-gradient(135deg, var(--success-color), var(--warning-color));
            border-radius: 15px;
            color: white;
            box-shadow: 0 5px 20px rgba(0, 0, 0, 0.1);
            transition: all 0.3s ease;
        }
        .stat-item:hover {
            transform: scale(1.05);
        }
        .stat-number {
            display: block;
            font-size: 2.5rem;
            font-weight: bold;
            margin-bottom: 0.5rem;
        }
        .stat-label {
            font-size: 0.9rem;
            opacity: 0.9;
        }
        .search-publish-section {
            display: flex;
            gap: 1rem;
            align-items: center;
            flex-wrap: wrap;
        }
        .search-input {
            flex: 1;
            min-width: 250px;
            padding: 1rem 1.5rem;
            border: 2px solid transparent;
            border-radius: 25px;
            background: rgba(255, 255, 255, 0.9);
            font-size: 1rem;
            transition: all 0.3s ease;
            box-shadow: 0 5px 15px rgba(0, 0, 0, 0.1);
        }
        .search-input:focus {
            outline: none;
            border-color: var(--primary-color);
            transform: translateY(-2px);
            box-shadow: 0 8px 25px rgba(0, 0, 0, 0.15);
        }
        .publish-button {
            background: linear-gradient(45deg, var(--primary-color), var(--secondary-color));
            color: white;
            border: none;
            padding: 1rem 2rem;
            border-radius: 25px;
            font-weight: 600;
            cursor: pointer;
            transition: all 0.3s ease;
            box-shadow: 0 5px 15px rgba(0, 0, 0, 0.2);
            white-space: nowrap;
        }
        .publish-button:hover {
            transform: translateY(-2px);
            box-shadow: 0 8px 25px rgba(0, 0, 0, 0.3);
        }
        .offer-item {
            background: rgba(255, 255, 255, 0.8);
            border-radius: 10px;
            padding: 1rem;
            margin-bottom: 1rem;
            box-shadow: 0 4px 10px rgba(0, 0, 0, 0.05);
            border-left: 4px solid var(--primary-color);
            transition: all 0.3s ease;
        }
        .offer-item:hover {
            transform: translateX(5px);
            box-shadow: 0 6px 15px rgba(0, 0, 0, 0.1);
        }
        .empty-state {
            text-align: center;
            padding: 3rem 2rem;
            color: #7f8c8d;
            background: rgba(255, 255, 255, 0.7);
            border-radius: 10px;
            margin: 1rem 0;
        }
        .empty-state .icon {
            font-size: 3rem;
            margin-bottom: 1rem;
            opacity: 0.6;
        }
        .quick-actions {
            list-style: none;
        }
        .quick-action-item {
            background: linear-gradient(135deg, rgba(102, 126, 234, 0.1), rgba(118, 75, 162, 0.1));
            border-radius: 12px;
            padding: 1rem 1.5rem;
            margin-bottom: 0.8rem;
            cursor: pointer;
            transition: all 0.3s ease;
            border-left: 3px solid var(--primary-color);
        }
        .quick-action-item:hover {
            background: linear-gradient(135deg, rgba(102, 126, 234, 0.2), rgba(118, 75, 162, 0.2));
            transform: translateX(5px);
        }
        .notification {
            position: fixed;
            top: 20px;
            right: 20px;
            background: linear-gradient(45deg, var(--success-color), var(--warning-color));
            color: white;
            padding: 1rem 1.5rem;
            border-radius: 10px;
            box-shadow: 0 5px 20px rgba(0, 0, 0, 0.2);
            transform: translateX(400px);
            opacity: 0;
            transition: all 0.3s ease;
            z-index: 1000;
        }
        .notification.show {
            transform: translateX(0);
            opacity: 1;
        }
        .progress-bar {
            width: 100%;
            height: 6px;
            background: rgba(255, 255, 255, 0.3);
            border-radius: 3px;
            overflow: hidden;
            margin-top: 1rem;
        }
        .progress-fill {
            height: 100%;
            background: linear-gradient(90deg, var(--accent-color), var(--success-color));
            width: 0%;
            transition: all 0.3s ease;
            border-radius: 3px;
        }
        @media (max-width: 768px) {
            .dashboard-content {
                grid-template-columns: 1fr;
                padding: 1rem;
            }
            .dashboard-header {
                padding: 1rem;
                flex-direction: column;
                gap: 1rem;
            }
            .stats-grid {
                grid-template-columns: 1fr;
            }
            .search-publish-section {
                flex-direction: column;
            }
            .search-input {
                min-width: auto;
                width: 100%;
            }
        }
        @keyframes fadeInUp {
            from {
                opacity: 0;
                transform: translateY(30px);
            }
            to {
                opacity: 1;
                transform: translateY(0);
            }
        }
        .dashboard-card {
            animation: fadeInUp 0.6s ease forwards;
        }
        .dashboard-card:nth-child(2) { animation-delay: 0.1s; }
        .dashboard-card:nth-child(3) { animation-delay: 0.2s; }
        .dashboard-card:nth-child(4) { animation-delay: 0.3s; }
        .loading {
            display: inline-block;
            width: 20px;
            height: 20px;
            border: 3px solid rgba(255, 255, 255, 0.3);
            border-radius: 50%;
            border-top-color: white;
            animation: spin 1s ease-in-out infinite;
        }
        @keyframes spin {
            to { transform: rotate(360deg); }
        }
        .btn {
            padding: 0.5rem 1rem;
            border-radius: 25px;
            font-weight: 500;
            transition: all 0.3s ease;
            cursor: pointer;
        }
        .btn-primary {
            background: linear-gradient(45deg, var(--primary-color), var(--secondary-color));
            color: white;
            border: none;
        }
        .btn-outline {
            background: transparent;
            color: var(--primary-color);
            border: 1px solid var(--primary-color);
        }
        .btn-small {
            padding: 0.3rem 0.8rem;
            font-size: 0.8rem;
        }
        .candidate-header {
            display: flex;
            justify-content: space-between;
            align-items: center;
            margin-bottom: 0.5rem;
        }
        .candidate-status {
            background: rgba(255, 255, 255, 0.3);
            padding: 0.2rem 0.5rem;
            border-radius: 10px;
            font-size: 0.8rem;
        }
        .candidate-details {
            margin-bottom: 1rem;
        }
        .candidate-actions {
            display: flex;
            gap: 0.5rem;
            justify-content: flex-end;
        }
        .project-meta {
            display: flex;
            justify-content: space-between;
            margin: 0.5rem 0;
            font-size: 0.9rem;
            color: #7f8c8d;
        }
        .project-actions {
            display: flex;
            justify-content: flex-end;
            gap: 0.5rem;
        }
        .search-publish-section form {
            display: flex;
            width: 100%;
            max-width: 600px;
        }
        .search-publish-section input[type="text"] {
            flex: 1;
            padding: 12px 20px;
            border: 2px solid #e2e8f0;
            border-radius: 30px 0 0 30px;
            font-size: 16px;
            outline: none;
            transition: border-color 0.3s ease, box-shadow 0.3s ease;
            box-shadow: 0 4px 6px rgba(0, 0, 0, 0.1);
        }
        .search-publish-section input[type="text"]:focus {
            border-color: #8052e6;
            box-shadow: 0 0 0 3px rgba(128, 82, 230, 0.2);
        }
        .search-publish-section button[type="submit"] {
            background: linear-gradient(135deg, #8052e6, #6a40d0);
            color: white;
            border: none;
            padding: 12px 24px;
            border-radius: 0 30px 30px 0;
            font-size: 16px;
            font-weight: bold;
            cursor: pointer;
            transition: transform 0.3s ease, box-shadow 0.3s ease;
            box-shadow: 0 4px 6px rgba(0, 0, 0, 0.1);
        }
        .search-publish-section button[type="submit"]:hover {
            transform: translateY(-2px);
            box-shadow: 0 6px 12px rgba(0, 0, 0, 0.2);
        }
        .search-publish-section input[type="text"]::placeholder {
            color: #aaa;
        }
        .header-actions {
            display: flex;
            align-items: center;
            gap: 1rem;
        }
        .notification-bell {
            font-size: 1.5rem;
            cursor: pointer;
            position: relative;
        }
        .notification-bell::after {
            content: attr(data-count);
            position: absolute;
            top: -5px;
            right: -5px;
            background-color: var(--danger-color);
            color: white;
            border-radius: 50%;
            width: 18px;
            height: 18px;
            display: flex;
            justify-content: center;
            align-items: center;
            font-size: 0.7rem;
        }
        .notification-panel {
            position: absolute;
            top: 70px;
            right: 20px;
            background: white;
            border-radius: 10px;
            box-shadow: 0 5px 20px rgba(0, 0, 0, 0.2);
            width: 300px;
            max-height: 400px;
            overflow-y: auto;
            display: none;
            z-index: 1001;
            padding: 1rem;
        }
        .notification-panel.show {
            display: block;
        }
        .notification-item {
            padding: 0.8rem;
            margin-bottom: 0.5rem;
            background: rgba(102, 126, 234, 0.1);
            border-radius: 8px;
            display: flex;
            align-items: center;
        }
        .notification-item:hover {
            background: rgba(102, 126, 234, 0.2);
        }
        .notification-item .icon {
            margin-right: 0.8rem;
            font-size: 1.2rem;
        }
        .notification-item .content {
            flex: 1;
        }
        .notification-item .content h4 {
            margin: 0 0 0.3rem 0;
            font-size: 0.9rem;
            color: var(--dark-color);
        }
        .notification-item .content p {
            margin: 0;
            font-size: 0.8rem;
            color: var(--text-color);
        }
        .notification-item .time {
            font-size: 0.7rem;
            color: #7f8c8d;
            margin-left: 0.5rem;
        }
        .carousel {
            position: relative;
            overflow: hidden;
            border-radius: 15px;
            box-shadow: 0 8px 20px rgba(0, 0, 0, 0.08);
            background: rgba(255, 255, 255, 0.95);
            padding: 1.5rem;
            margin-bottom: 1.5rem;
        }
        .carousel-container {
            display: flex;
            transition: transform 0.5s ease;
        }
        .carousel-item {
            min-width: 100%;
            padding: 1rem;
        }
        .carousel-item h3 {
            margin-bottom: 0.5rem;
            color: var(--dark-color);
        }
        .carousel-item p {
            color: var(--text-color);
        }
        .carousel-controls {
            position: absolute;
            top: 50%;
            width: 100%;
            display: flex;
            justify-content: space-between;
            transform: translateY(-50%);
        }
        .carousel-control {
            background: rgba(255, 255, 255, 0.5);
            border: none;
            color: var(--dark-color);
            padding: 0.5rem;
            cursor: pointer;
            border-radius: 50%;
            width: 40px;
            height: 40px;
            display: flex;
            align-items: center;
            justify-content: center;
            box-shadow: 0 2px 5px rgba(0, 0, 0, 0.2);
        }
        .carousel-control:hover {
            background: rgba(255, 255, 255, 0.8);
        }
    </style>
</head>
<body>
    <div class="dashboard" id="recruiterDashboard">
        <div class="dashboard-header">
            <h1>🏢 Tableau de bord Recruteur</h1>
            <div class="header-actions">
                <div class="notification-bell" id="notificationBell" data-count="0">🔔</div>
                <button class="cta-button" onclick="window.location.href='/candidats_projets'">👥 Voir les candidats et projets</button>

                <button id="logoutBtnRecruiter" class="cta-button" onclick="window.location.href='/logout'">Déconnexion</button>
            </div>
            <div class="notification-panel" id="notificationPanel">
                <!-- Les notifications seront ajoutées ici -->
            </div>
        </div>
        <div class="dashboard-content">
            <div class="main-content">
                <div class="dashboard-card">
                    <h2>📊 Vue d'ensemble</h2>
                    <div class="stats-grid">
                        <div class="stat-item">
                            <span class="stat-number" id="activeOffers">0</span>
                            <span class="stat-label">Offres actives</span>
                        </div>
                    </div>
                    <div class="search-publish-section">
                        <form action="/recherche-talents" method="get">
                            <input type="text" name="q" placeholder="🔍 Rechercher des talents...">
                            <button type="submit">Rechercher</button>
                        </form>
                    </div>
                </div>
                <div class="dashboard-card">
                    <h2>📋 Offres publiées</h2>
                    <div id="offersContainer">
                        <div class="empty-state">
                            <div class="icon">📄</div>
                            <p>Aucune offre publiée</p>
                            <p>Commencez par publier votre première offre d'emploi</p>
                            <button class="publish-button" id="publishOfferBtn2">Publier une offre</button>
                        </div>
                    </div>
                </div>
                <div class="carousel">
                    <div class="carousel-container" id="carouselContainer">
                        <div class="carousel-item">
                            <h3>💬 Conseil 1 – Soyez réactif aux candidatures</h3>
                            <p>⚡ Répondez rapidement aux nouvelles candidatures pour ne pas perdre les meilleurs talents.</p>
                        </div>
                        <div class="carousel-item">
                            <h3>📈 Conseil 2 – Publiez des offres claires et détaillées</h3>
                            <p>✍️ Décrivez précisément la mission, les compétences requises et le format (remote, stage, etc.) pour un meilleur matching.</p>
                        </div>
                        <div class="carousel-item">
                            <h3>🎯 Conseil 3 – Utilisez le matching IA</h3>
                            <p>🤖 Laissez l’algorithme filtrer pour vous les meilleurs profils compatibles avec votre offre.</p>
                        </div>
                        <div class="carousel-item">
                            <h3>👀 Conseil 4 – Consultez les CV depuis les notifications</h3>
                            <p>📎 Chaque notification inclut un lien direct vers le CV : prenez 30 secondes pour jeter un œil avant de refuser.</p>
                        </div>
                        <div class="carousel-item">
                         – Soyez bienveillant dans les réponses</h3>
                            <p>💬 Même en cas de refus, un retour respectueux améliore votre image de recruteur.</p>
                        </div>
                    </div>
                    <div class="carousel-controls">
                        <button class="carousel-control" id="prevBtn">❮</button>
                        <button class="carousel-control" id="nextBtn">❯</button>
                    </div>
                </div>
            </div>
            <div class="sidebar">
                <div class="dashboard-card">
                    <h2>⚡ Actions rapides</h2>
                    <ul class="quick-actions">
                        <li class="quick-action-item" onclick="window.location.href='/recruiter/scrape_remoteok?q=python'">
                        🌐 Importer des offres depuis d'autres sites
                        </li>
                        <li class="quick-action-item" id="newOfferAction">
                            📝 Publier une nouvelle offre
                        </li>
                        <li class="quick-action-item" id="completeProfileAction">
                            🏢 Compléter le profil entreprise
                        </li>
                        <li class="quick-action-item" id="messagesAction" onclick="window.location.href='/messages'">
                            💬 Consulter les messages
                        </li>
                       <li class="quick-action-item" onclick="window.location.href='/recruiter/stats_report'">
                            📈 Rapport visuel complet 📊
                        </li>
                        <li class="quick-action-item" id="settingsAction" onclick="window.location.href='/recruteur/settings'">
                            ⚙️ Paramètres du compte
                        </li>
                    </ul>
                </div>
            </div>
        </div>
    </div>
    <div class="notification" id="notification"></div>
    <script>
        class RecruiterDashboard {
            constructor() {
                this.data = {
                    offers: [],
                    stats: {
                        activeOffers: 0,
                        totalApplications: 0,
                        activeProjects: 0
                    }
                };
                this.init();
            }
            init() {
                this.bindEvents();
                this.loadData();
                this.updateStats();
                this.startAutoRefresh();
                this.initCarousel();
            }
            bindEvents() {
                document.getElementById('publishOfferBtn')?.addEventListener('click', () => this.publishOffer());
                document.getElementById('publishOfferBtn2')?.addEventListener('click', () => this.publishOffer());
                document.getElementById('newOfferAction')?.addEventListener('click', () => this.publishOffer());
                document.getElementById('completeProfileAction')?.addEventListener('click', () => this.completeProfile());
                document.getElementById('messagesAction')?.addEventListener('click', () => this.openMessages());
                document.getElementById('analyticsAction')?.addEventListener('click', () => this.showAnalytics());
                document.getElementById('settingsAction')?.addEventListener('click', () => this.openSettings());
            }
            async loadData() {
                try {
                    const offersResponse = await fetch('/api/offers');
                    const offersData = await offersResponse.json();
                    this.data.offers = offersData.offers.map(o => ({
                        id: o._id,
                        title: o.title,
                        description: o.description,
                        status: o.status,
                        applications: 0,
                        createdAt: new Date()
                    }));
                    this.updateStats();
                    this.renderOffers();
                } catch (error) {
                    console.error('Error loading offers:', error);
                }
            }
            updateStats() {
                this.data.stats.activeOffers = this.data.offers.filter(o => o.status === 'active').length;
                this.data.stats.totalApplications = this.data.offers.reduce((sum, offer) => sum + offer.applications, 0);
                this.animateCounter('activeOffers', this.data.stats.activeOffers);
                this.animateCounter('totalApplications', this.data.stats.totalApplications);
            }
            animateCounter(elementId, targetValue) {
                const element = document.getElementById(elementId);
                if (!element) return;
                let current = 0;
                const increment = targetValue / 30;
                const timer = setInterval(() => {
                    current += increment;
                    if (current >= targetValue) {
                        current = targetValue;
                        clearInterval(timer);
                    }
                    element.textContent = Math.floor(current);
                }, 50);
            }
            renderOffers() {
                const container = document.getElementById('offersContainer');
                if (!container) return;
                if (this.data.offers.length === 0) {
                    container.innerHTML = `
                        <div class="empty-state">
                            <div class="icon">📄</div>
                            <p>Aucune offre publiée</p>
                            <p>Commencez par publier votre première offre d'emploi</p>
                            <button class="publish-button" onclick="dashboard.publishOffer()">Publier une offre</button>
                        </div>
                    `;
                    return;
                }
                container.innerHTML = this.data.offers.map(offer => `
                    <div class="offer-item">
                        <h3>${offer.title}</h3>
                        <p><strong>Description:</strong> ${offer.description}</p>
                        <p><strong>Candidatures:</strong> ${offer.applications}</p>
                        <p><strong>Statut:</strong> <span style="color: #27ae60;">${offer.status === 'active' ? 'Active' : 'Inactive'}</span></p>
                        <p><strong>Publiée le:</strong> ${offer.createdAt.toLocaleDateString()}</p>
                        <div style="margin-top: 1rem;">
                            <button class="btn btn-small btn-outline" onclick="window.location.href='/edit_offer/${offer.id}'">Modifier</button>
                            <button class="btn btn-small btn-primary" onclick="window.location.href='/delete_offer/${offer.id}'">Supprimer</button>
                        </div>
                    </div>
                `).join('');
            }

            publishOffer() {
                window.location.href = '/create_offer';
            }
            completeProfile() {
                window.location.href = "/complete_profile";
            }
            openMessages() {
                window.location.href = '/messages';
            }
            showAnalytics() {
                window.location.href = '/recruteur/statistiques';
            }
            openSettings() {
                window.location.href = '/recruteur/settings';
            }
            showNotification(message) {
                const notification = document.getElementById('notification');
                if (notification) {
                    notification.textContent = message;
                    notification.classList.add('show');
                    setTimeout(() => {
                        notification.classList.remove('show');
                    }, 3000);
                }
            }
            startAutoRefresh() {
                setInterval(() => {
                    if (Math.random() > 0.7) {
                        this.data.stats.totalApplications += Math.floor(Math.random() * 3);
                        this.updateStats();
                    }
                }, 30000);
            }
            initCarousel() {
                const container = document.getElementById('carouselContainer');
                const items = document.querySelectorAll('.carousel-item');
                const prevBtn = document.getElementById('prevBtn');
                const nextBtn = document.getElementById('nextBtn');
                let currentIndex = 0;
                const itemWidth = items[0].clientWidth;
                let interval = setInterval(nextSlide, 5000);

                function updateCarousel() {
                    container.style.transform = `translateX(-${currentIndex * itemWidth}px)`;
                }

                function nextSlide() {
                    currentIndex = (currentIndex + 1) % items.length;
                    updateCarousel();
                }

                function prevSlide() {
                    currentIndex = (currentIndex - 1 + items.length) % items.length;
                    updateCarousel();
                }

                nextBtn.addEventListener('click', () => {
                    clearInterval(interval);
                    nextSlide();
                    interval = setInterval(nextSlide, 5000);
                });

                prevBtn.addEventListener('click', () => {
                    clearInterval(interval);
                    prevSlide();
                    interval = setInterval(nextSlide, 5000);
                });

                updateCarousel();
            }
        }
        const dashboard = new RecruiterDashboard();
        document.addEventListener('DOMContentLoaded', () => {
            document.body.style.opacity = '0';
            setTimeout(() => {
                document.body.style.transition = 'opacity 0.5s ease';
                document.body.style.opacity = '1';
            }, 100);
        });
        document.addEventListener('DOMContentLoaded', function() {
            fetch('/recruteur/stats')
                .then(response => response.json())
                .then(data => {
                    document.getElementById('activeOffers').textContent = data.offres;
                    document.getElementById('totalApplications').textContent = data.candidatures;
                });
        });

       document.getElementById("notificationBell").addEventListener("click", function () {
    const panel = document.getElementById("notificationPanel");
    panel.classList.toggle("show");
    fetch("/api/get_notifications")
    .then(res => res.json())
    .then(data => {
        panel.innerHTML = data.map(n => `
            <div class="notification-item">
                <div class="icon">🔔</div>
                <div class="content">
                    <h4>${n.message}</h4>
                    <p class="time">${n.timestamp}</p>
                </div>
            </div>
        `).join('');
        document.getElementById("notificationBell").setAttribute("data-count", data.length);
    });
});
    </script>
</body>
</html>


"""
recruiter_scraped_offers="""
<!DOCTYPE html>
<html lang="fr">
<head>
    <meta charset="UTF-8">
    <title>Offres Scrapées - Web4Jobs</title>
    <style>
        body {
            font-family: Arial, sans-serif;
            background: #f7f7f7;
            padding: 2rem;
            color: #333;
        }
        h2 {
            text-align: center;
            margin-bottom: 2rem;
            color: #4a148c;
        }
        .offer-card {
            background: #fff;
            border-radius: 10px;
            padding: 1.5rem;
            margin-bottom: 1.5rem;
            box-shadow: 0 2px 6px rgba(0,0,0,0.1);
            border-left: 5px solid #6a1b9a;
        }
        .offer-card h3 {
            margin-top: 0;
            color: #6a1b9a;
        }
        .offer-card p {
            margin: 0.5rem 0;
        }
        form {
            margin-top: 1rem;
        }
        button {
            background-color: #6a1b9a;
            color: white;
            border: none;
            padding: 0.6rem 1.2rem;
            border-radius: 6px;
            cursor: pointer;
        }
        button:hover {
            background-color: #4a148c;
        }
    </style>
</head>
<body>

    <h2>🌐 Importer des offres depuis d'autres sites
</h2>

    {% for offer in offers %}
    <div class="offer-card">
        <h3>{{ offer.title }}</h3>
        <p><strong>Entreprise :</strong> {{ offer.company }}</p>
        <p><strong>Lieu :</strong> {{ offer.location }}</p>
        <p>{{ offer.description }}</p>
        <form method="POST" action="/recruiter/publish_scraped_offer">
            <input type="hidden" name="title" value="{{ offer.title }}">
            <input type="hidden" name="description" value="{{ offer.description }}">
            <input type="hidden" name="tags" value="{{ offer.tags | join(',') }}">
            <button type="submit">📤 Publier sur Web4Jobs</button>
        </form>
    </div>
    {% endfor %}

</body>
</html>

"""


ADMIN_DASHBOARD_TEMPLATE = """
<!DOCTYPE html>
<html lang="fr">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Tableau de bord Administrateur</title>
    <style>
        :root {
            --primary-color: #8052e6;
            --primary-dark: #6a40d0;
            --secondary-color: #2c2c6c;
            --accent-color: #ff6b6b;
            --success-color: #4ecdc4;
            --warning-color: #ffe66d;
            --background-color: #f8fafc;
            --card-background: #ffffff;
            --text-primary: #2d3748;
            --text-secondary: #718096;
            --border-color: #e2e8f0;
            --shadow-light: 0 4px 6px rgba(0, 0, 0, 0.07);
            --shadow-medium: 0 10px 25px rgba(0, 0, 0, 0.1);
            --shadow-heavy: 0 20px 40px rgba(0, 0, 0, 0.15);
            --border-radius: 12px;
            --transition: all 0.3s cubic-bezier(0.4, 0, 0.2, 1);
        }

        * {
            margin: 0;
            padding: 0;
            box-sizing: border-box;
        }

        body {
            font-family: 'Segoe UI', sans-serif;
            background: var(--background-color);
            color: var(--text-primary);
            line-height: 1.6;
            padding: 0;
            margin: 0;
            display: flex;
            min-height: 100vh;
            overflow-x: hidden;
        }

        .dashboard {
            display: flex;
            width: 100%;
            transition: var(--transition);
        }

        .sidebar {
            width: 250px;
            background: linear-gradient(135deg, var(--primary-color), var(--primary-dark));
            color: white;
            padding: 1.5rem;
            height: 100vh;
            position: fixed;
            box-shadow: var(--shadow-medium);
            transition: var(--transition);
            transform: translateX(0);
            z-index: 1000;
        }

        .sidebar.collapsed {
            transform: translateX(-200px);
            width: 50px;
            overflow: hidden;
        }

        .sidebar-header {
            margin-bottom: 2rem;
            text-align: center;
            display: flex;
            justify-content: space-between;
            align-items: center;
        }

        .sidebar-header h2 {
            font-size: 1.5rem;
            margin-bottom: 0.5rem;
        }

        .toggle-btn {
            background: none;
            border: none;
            color: white;
            font-size: 1.5rem;
            cursor: pointer;
            padding: 0.5rem;
            transition: var(--transition);
        }

        .toggle-btn.collapsed {
            transform: rotate(180deg);
        }

        .sidebar-menu {
            list-style: none;
        }

        .sidebar-menu li {
            margin-bottom: 1rem;
        }

        .sidebar-menu li a {
            color: white;
            text-decoration: none;
            display: flex;
            align-items: center;
            padding: 0.75rem 1rem;
            border-radius: var(--border-radius);
            transition: var(--transition);
        }

        .sidebar-menu li a:hover {
            background: rgba(255, 255, 255, 0.1);
        }

        .sidebar-menu li a i {
            margin-right: 0.75rem;
            font-size: 1.2rem;
        }

        .main-content {
            margin-left: 250px;
            width: calc(100% - 250px);
            padding: 2rem;
            transition: var(--transition);
        }

        .main-content.expanded {
            margin-left: 50px;
            width: calc(100% - 50px);
        }

        .overlay {
            position: fixed;
            top: 0;
            left: 0;
            width: 100%;
            height: 100%;
            background-color: rgba(0, 0, 0, 0.5);
            z-index: 999;
            opacity: 0;
            pointer-events: none;
            transition: var(--transition);
        }

        .overlay.active {
            opacity: 1;
            pointer-events: all;
        }

        .dashboard-header {
            background: linear-gradient(135deg, var(--secondary-color), var(--primary-color));
            color: white;
            padding: 2rem;
            margin-bottom: 2rem;
            border-radius: var(--border-radius);
            box-shadow: var(--shadow-medium);
            display: flex;
            justify-content: space-between;
            align-items: center;
        }

        .dashboard-header h1 {
            font-size: 1.8rem;
            font-weight: 600;
        }

        .dashboard-content {
            display: grid;
            grid-template-columns: 1fr;
            gap: 2rem;
        }

        .dashboard-card {
            background: var(--card-background);
            padding: 2rem;
            border-radius: var(--border-radius);
            box-shadow: var(--shadow-light);
            border: 1px solid var(--border-color);
            transition: var(--transition);
        }

        .dashboard-card:hover {
            transform: translateY(-5px);
            box-shadow: var(--shadow-medium);
        }

        .dashboard-card h2 {
            color: var(--primary-color);
            margin-bottom: 1rem;
            display: flex;
            align-items: center;
            gap: 0.5rem;
        }

        .stats-grid {
            display: grid;
            grid-template-columns: repeat(auto-fit, minmax(200px, 1fr));
            gap: 1.5rem;
            margin: 1rem 0;
        }

        .stat-item {
            text-align: center;
            padding: 1.5rem;
            background: rgba(128, 82, 230, 0.05);
            border-radius: var(--border-radius);
            transition: var(--transition);
            box-shadow: var(--shadow-light);
        }

        .stat-item:hover {
            background: rgba(128, 82, 230, 0.1);
            transform: scale(1.05);
        }

        .stat-number {
            font-size: 2.5rem;
            font-weight: 700;
            color: var(--primary-color);
            display: block;
            margin-bottom: 0.5rem;
        }

        .stat-label {
            font-size: 1rem;
            color: var(--text-secondary);
            margin-top: 0.5rem;
            font-weight: 500;
        }

        .user-table {
            width: 100%;
            border-collapse: collapse;
            margin: 1rem 0;
        }

        .user-table th, .user-table td {
            padding: 0.75rem;
            text-align: left;
            border-bottom: 1px solid var(--border-color);
        }

        .user-table th {
            background-color: rgba(128, 82, 230, 0.1);
            font-weight: bold;
        }

        .user-table tr:hover {
            background-color: rgba(128, 82, 230, 0.05);
        }

        .user-actions {
            display: flex;
            gap: 0.5rem;
        }

        .user-actions button {
            background: linear-gradient(135deg, var(--primary-color), var(--primary-dark));
            color: white;
            border: none;
            padding: 0.5rem 1rem;
            border-radius: var(--border-radius);
            cursor: pointer;
            transition: var(--transition);
        }

        .user-actions button:hover {
            transform: translateY(-2px);
            box-shadow: var(--shadow-medium);
        }

        .action-button {
            background: linear-gradient(135deg, var(--primary-color), var(--primary-dark));
            color: white;
            border: none;
            padding: 0.75rem 1.5rem;
            border-radius: var(--border-radius);
            font-weight: 600;
            cursor: pointer;
            transition: var(--transition);
            margin-top: 1rem;
            display: inline-block;
        }

        .action-button:hover {
            transform: translateY(-2px);
            box-shadow: var(--shadow-medium);
        }

        @media (max-width: 768px) {
            .sidebar {
                transform: translateX(-250px);
            }

            .sidebar.active {
                transform: translateX(0);
            }

            .main-content {
                margin-left: 0;
                width: 100%;
            }

            .main-content.expanded {
                margin-left: 0;
                width: 100%;
            }

            .overlay.active {
                display: block;
            }

            .dashboard-content {
                grid-template-columns: 1fr;
            }

            .dashboard-header {
                flex-direction: column;
                text-align: center;
                gap: 1rem;
            }

            .stats-grid {
                grid-template-columns: 1fr;
            }
        }

        @keyframes fadeInUp {
            from {
                opacity: 0;
                transform: translateY(30px);
            }
            to {
                opacity: 1;
                transform: translateY(0);
            }
        }
    </style>
</head>
<body>
    <div class="dashboard">
        <div class="sidebar" id="sidebar">
            <div class="sidebar-header">
                <h2>⚙️ Admin</h2>
                <button class="toggle-btn" onclick="toggleSidebar()">❮</button>
            </div>
            <ul class="sidebar-menu">
                <li><a href="/dashboard/admin"><i>📊</i><span class="menu-text"> Statistiques</span></a></li>
                <li><a href="/generate_reports"><i>📈</i><span class="menu-text"> Rapports</span></a></li>
                <li><a href="/messages"><i>💬</i><span class="menu-text"> Messagerie</span></a></li>
                <li><a href="/settings"><i>⚙️</i><span class="menu-text"> Paramètres</span></a></li>
                <li><a href="/logout"><i>🚪</i><span class="menu-text"> Déconnexion</span></a></li>
            </ul>
        </div>

        <div class="main-content" id="mainContent">
            <div class="dashboard-header">
                <h1>📊 Statistiques Globales</h1>
            </div>

            <div class="dashboard-content">
                <!-- Section Vue d'ensemble -->
                <div class="dashboard-card">
                    <h2>📊 Vue d'ensemble</h2>
                    <div class="stats-grid">
                        <div class="stat-item">
                            <span class="stat-number" id="totalUsers">{{ total_users }}</span>
                            <span class="stat-label">Utilisateurs</span>
                        </div>
                        <div class="stat-item">
                            <span class="stat-number" id="totalOffers">{{ total_offers }}</span>
                            <span class="stat-label">Offres</span>
                        </div>
                        <div class="stat-item">
                            <span class="stat-number" id="totalApplications">{{ total_applications }}</span>
                            <span class="stat-label">Candidatures</span>
                        </div>
                    </div>
                </div>

                <!-- Section Gestion des Utilisateurs -->
                <div class="dashboard-card">
                    <h2>📋 Gestion des Utilisateurs</h2>
                    <div class="user-table-container">
                        <table class="user-table">
                            <thead>
                                <tr>
                                    <th>Nom</th>
                                    <th>Email</th>
                                    <th>Type</th>
                                    <th>Actions</th>
                                </tr>
                            </thead>
                            <tbody>
                                {% for user in users %}
                                <tr data-user-id="{{ user._id }}">
    <td>{{ user.full_name }}</td>
    <td>{{ user.email }}</td>
    <td>{{ user.user_type }}</td>
    <td class="user-actions">
        <button onclick="window.location.href='/edit_user/{{ user._id }}'">Modifier</button>
        <button onclick="deleteUser('{{ user._id }}')">Supprimer</button>
    </td>
</tr>
                                {% endfor %}
                            </tbody>
                        </table>
                        <button class="action-button" id="addUserBtn" onclick="window.location.href='/add_user'">Ajouter un utilisateur</button>
                    </div>
                </div>
            </div>
        </div>
    </div>

    <div class="overlay" id="overlay"></div>

    <script>
        function toggleSidebar() {
            const sidebar = document.getElementById('sidebar');
            const mainContent = document.getElementById('mainContent');
            const overlay = document.getElementById('overlay');
            const toggleBtn = document.querySelector('.toggle-btn');

            sidebar.classList.toggle('collapsed');
            mainContent.classList.toggle('expanded');
            toggleBtn.classList.toggle('collapsed');

            if (window.innerWidth <= 768) {
                sidebar.classList.toggle('active');
                overlay.classList.toggle('active');
            }
        }

        // Fermer la sidebar lorsque l'overlay est cliqué sur mobile
        document.getElementById('overlay').addEventListener('click', function() {
            const sidebar = document.getElementById('sidebar');
            const mainContent = document.getElementById('mainContent');
            const overlay = document.getElementById('overlay');
            const toggleBtn = document.querySelector('.toggle-btn');

            sidebar.classList.remove('active');
            overlay.classList.remove('active');
            sidebar.classList.add('collapsed');
            mainContent.classList.add('expanded');
            toggleBtn.classList.add('collapsed');
        });

        // Gérer le redimensionnement de la fenêtre
        window.addEventListener('resize', function() {
            const sidebar = document.getElementById('sidebar');
            const mainContent = document.getElementById('mainContent');
            const overlay = document.getElementById('overlay');
            const toggleBtn = document.querySelector('.toggle-btn');

            if (window.innerWidth > 768) {
                sidebar.classList.remove('active');
                overlay.classList.remove('active');
                // Par défaut, la sidebar est ouverte sur les grands écrans
                sidebar.classList.remove('collapsed');
                mainContent.classList.remove('expanded');
                toggleBtn.classList.remove('collapsed');
            } else {
                // Sur mobile, la sidebar est fermée par défaut
                sidebar.classList.add('collapsed');
                mainContent.classList.add('expanded');
                toggleBtn.classList.add('collapsed');
            }
        });

        // Initialisation au chargement de la page
        window.addEventListener('load', function() {
            const sidebar = document.getElementById('sidebar');
            const mainContent = document.getElementById('mainContent');
            const toggleBtn = document.querySelector('.toggle-btn');

            if (window.innerWidth <= 768) {
                sidebar.classList.add('collapsed');
                mainContent.classList.add('expanded');
                toggleBtn.classList.add('collapsed');
            }
        });

        function deleteUser(userId) {
    if (confirm('Êtes-vous sûr de vouloir supprimer cet utilisateur ?')) {
        fetch('/delete_user/' + userId, {
            method: 'DELETE'
        })
        .then(response => response.json())
        .then(data => {
            if (data.success) {
                // Supprimer la ligne HTML sans recharger
                const row = document.querySelector(`[data-user-id="${userId}"]`);
                if (row) row.remove();
                alert("✅ Utilisateur supprimé !");
            } else {
                alert('❌ Échec : ' + data.message);
            }
        })
        .catch(error => {
            console.error('Erreur JS :', error);
            alert('❌ Erreur réseau.');
        });
    }
}
    </script>
</body>
</html>



"""

COMPLETE_PROFILE_TEMPLATE = """
<!DOCTYPE html>
<html lang="fr">
<head>
  <meta charset="UTF-8">
  <title>Profil Entreprise</title>
  <style>
    :root {
      --primary-color: #8052e6;
      --primary-dark: #6a40d0;
      --secondary-color: #2c2c6c;
      --accent-color: #ff6b6b;
      --success-color: #4ecdc4;
      --warning-color: #ffe66d;
      --background-color: #f8fafc;
      --card-background: #ffffff;
      --text-primary: #2d3748;
      --text-secondary: #718096;
      --border-color: #e2e8f0;
      --shadow-light: 0 4px 6px rgba(0, 0, 0, 0.07);
      --shadow-medium: 0 10px 25px rgba(0, 0, 0, 0.1);
      --shadow-heavy: 0 20px 40px rgba(0, 0, 0, 0.15);
      --border-radius: 12px;
      --transition: all 0.3s cubic-bezier(0.4, 0, 0.2, 1);
    }
    * {
      margin: 0;
      padding: 0;
      box-sizing: border-box;
    }
    body {
      font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
      background: var(--background-color);
      color: var(--text-primary);
      line-height: 1.6;
      padding: 2rem;
    }
    .container {
      max-width: 600px;
      margin: auto;
      background: var(--card-background);
      padding: 2rem;
      border-radius: var(--border-radius);
      box-shadow: var(--shadow-medium);
      border: 1px solid var(--border-color);
    }
    h2 {
      text-align: center;
      color: var(--primary-color);
      margin-bottom: 1.5rem;
    }
    .form-group {
      margin-bottom: 1.5rem;
    }
    .form-group label {
      display: block;
      margin-bottom: 0.5rem;
      font-weight: 600;
      color: var(--text-primary);
    }
    .form-group input,
    .form-group textarea {
      width: 100%;
      padding: 1rem;
      border: 2px solid var(--border-color);
      border-radius: var(--border-radius);
      font-size: 1rem;
      transition: var(--transition);
    }
    .form-group input:focus,
    .form-group textarea:focus {
      outline: none;
      border-color: var(--primary-color);
      box-shadow: 0 0 0 3px rgba(128, 82, 230, 0.1);
    }
    .form-group textarea {
      min-height: 150px;
      resize: vertical;
    }
    .submit-button {
      background: linear-gradient(135deg, var(--primary-color), var(--primary-dark));
      color: white;
      border: none;
      padding: 1rem;
      border-radius: var(--border-radius);
      font-size: 1rem;
      font-weight: 600;
      cursor: pointer;
      transition: var(--transition);
      display: block;
      width: 100%;
      margin-top: 1rem;
    }
    .submit-button:hover {
      transform: translateY(-2px);
      box-shadow: var(--shadow-medium);
    }
    .notification {
      position: fixed;
      top: 20px;
      right: 20px;
      background: linear-gradient(45deg, var(--success-color), var(--warning-color));
      color: white;
      padding: 1rem 1.5rem;
      border-radius: 10px;
      box-shadow: 0 5px 20px rgba(0, 0, 0, 0.2);
      transform: translateX(400px);
      opacity: 0;
      transition: all 0.3s ease;
      z-index: 1000;
    }
    .notification.show {
      transform: translateX(0);
      opacity: 1;
    }
  </style>
</head>
<body>
  <div class="container">
    <h2>Compléter le profil entreprise</h2>
    <form id="companyProfileForm" method="POST">
      <div class="form-group">
        <label for="company_name">Nom de l’entreprise</label>
        <input type="text" id="company_name" name="company_name" required>
      </div>
      <div class="form-group">
        <label for="sector">Secteur d’activité</label>
        <input type="text" id="sector" name="sector" required>
      </div>
      <div class="form-group">
        <label for="size">Taille de l’entreprise</label>
        <input type="text" id="size" name="size">
      </div>
      <div class="form-group">
        <label for="description">Description</label>
        <textarea id="description" name="description" rows="5"></textarea>
      </div>
      <div class="form-group">
        <label for="logo_url">Logo (URL)</label>
        <input type="text" id="logo_url" name="logo_url">
      </div>
      <button type="submit" class="submit-button">Enregistrer</button>
    </form>
  </div>

  <div class="notification" id="notification"></div>

  <script>
    document.getElementById('companyProfileForm').addEventListener('submit', function(event) {
      let isValid = true;
      const inputs = document.querySelectorAll('input[required], textarea[required]');

      inputs.forEach(input => {
        if (!input.value.trim()) {
          isValid = false;
          input.style.borderColor = 'var(--accent-color)';
        } else {
          input.style.borderColor = 'var(--border-color)';
        }
      });

      if (!isValid) {
        event.preventDefault();
        showNotification('Veuillez remplir tous les champs obligatoires.');
      } else {
        showNotification('Profil enregistré avec succès!');
      }
    });

    function showNotification(message) {
      const notification = document.getElementById('notification');
      notification.textContent = message;
      notification.classList.add('show');

      setTimeout(() => {
        notification.classList.remove('show');
      }, 3000);
    }
  </script>
</body>
</html>
"""

MESSAGE_TEMPLATE = """
<!DOCTYPE html>
<html lang="fr">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Messagerie - Web4.Jobs</title>
    <style>
        :root {
            --primary-color: #8052e6;
            --primary-dark: #6a40d0;
            --secondary-color: #2c2c6c;
            --accent-color: #ff6b6b;
            --success-color: #4ecdc4;
            --warning-color: #ffe66d;
            --background-color: #f8fafc;
            --card-background: #ffffff;
            --text-primary: #2d3748;
            --text-secondary: #718096;
            --border-color: #e2e8f0;
            --shadow-light: 0 4px 6px rgba(0, 0, 0, 0.07);
            --shadow-medium: 0 10px 25px rgba(0, 0, 0, 0.1);
            --shadow-heavy: 0 20px 40px rgba(0, 0, 0, 0.15);
            --border-radius: 12px;
            --transition: all 0.3s cubic-bezier(0.4, 0, 0.2, 1);
        }
        * {
            margin: 0;
            padding: 0;
            box-sizing: border-box;
        }
        body {
            font-family: 'Segoe UI', sans-serif;
            background: var(--background-color);
            color: var(--text-primary);
            line-height: 1.6;
            padding: 2rem;
            min-height: 100vh;
            display: flex;
            justify-content: center;
            align-items: center;
            background-image: linear-gradient(135deg, rgba(128, 82, 230, 0.1), rgba(255, 255, 255, 0.9));
        }
        .messagerie {
            max-width: 800px;
            width: 100%;
            background-color: var(--card-background);
            border-radius: var(--border-radius);
            padding: 2rem;
            box-shadow: var(--shadow-heavy);
            border: 1px solid var(--border-color);
            transition: var(--transition);
            animation: fadeInUp 0.8s ease-out;
        }
        h2 {
            text-align: center;
            color: var(--primary-color);
            margin-bottom: 1.5rem;
            font-size: 2rem;
        }
        .form-group {
            margin-bottom: 1.5rem;
        }
        .form-group label {
            display: block;
            margin-bottom: 0.5rem;
            font-weight: 600;
            color: var(--text-primary);
        }
        input, textarea, select {
            width: 100%;
            padding: 1rem;
            border: 2px solid var(--border-color);
            border-radius: var(--border-radius);
            font-size: 1rem;
            transition: var(--transition);
            margin-bottom: 1rem;
        }
        input:focus, textarea:focus, select:focus {
            outline: none;
            border-color: var(--primary-color);
            box-shadow: 0 0 0 3px rgba(128, 82, 230, 0.1);
        }
        button {
            background: linear-gradient(135deg, var(--primary-color), var(--primary-dark));
            color: white;
            border: none;
            padding: 1rem;
            border-radius: var(--border-radius);
            font-size: 1rem;
            font-weight: 600;
            cursor: pointer;
            transition: var(--transition);
            width: 100%;
        }
        button:hover {
            transform: translateY(-2px);
            box-shadow: var(--shadow-medium);
        }
        .message {
            padding: 1.5rem;
            margin-bottom: 1rem;
            border-radius: var(--border-radius);
            transition: var(--transition);
            position: relative;
            overflow: hidden;
        }
        .sent {
            background-color: var(--success-color);
            text-align: right;
            margin-left: 2rem;
            animation: slideInRight 0.5s ease-out;
        }
        .received {
            background-color: var(--warning-color);
            text-align: left;
            margin-right: 2rem;
            animation: slideInLeft 0.5s ease-out;
        }
        .timestamp {
            font-size: 0.8rem;
            color: var(--text-secondary);
            margin-top: 0.5rem;
        }
        .message-actions {
            display: flex;
            justify-content: flex-end;
            margin-top: 0.5rem;
        }
        .message-actions button {
            background: none;
            border: none;
            color: var(--text-secondary);
            cursor: pointer;
            padding: 0.25rem 0.5rem;
            border-radius: 4px;
            margin-left: 0.5rem;
        }
        .message-actions button:hover {
            background: rgba(0, 0, 0, 0.1);
        }
        hr {
            margin: 1.5rem 0;
            border: 0;
            height: 1px;
            background: linear-gradient(90deg, transparent, var(--border-color), transparent);
        }
        @keyframes fadeInUp {
            from {
                opacity: 0;
                transform: translateY(30px);
            }
            to {
                opacity: 1;
                transform: translateY(0);
            }
        }
        @keyframes slideInLeft {
            from {
                opacity: 0;
                transform: translateX(-50px);
            }
            to {
                opacity: 1;
                transform: translateX(0);
            }
        }
        @keyframes slideInRight {
            from {
                opacity: 0;
                transform: translateX(50px);
            }
            to {
                opacity: 1;
                transform: translateX(0);
            }
        }
    </style>
</head>
<body>
    <div class="messagerie">
        <h2>💬 Messagerie</h2>

        <form method="get">
            <label for="user">Filtrer par utilisateur :</label>
            <select name="user" onchange="this.form.submit()">
                <option value="">-- Voir toutes les conversations --</option>
                {% for user in all_users %}
                    {% if user.email != current_user %}
                        <option value="{{ user.email }}" {% if user.email == selected_user %}selected{% endif %}>{{ user.email }}</option>
                    {% endif %}
                {% endfor %}
            </select>
        </form>

        <form method="post">
            <label for="recipient">À :</label>
            <input list="userlist" name="recipient" required placeholder="Email du destinataire">
            <datalist id="userlist">
                {% for user in all_users %}
                    <option value="{{ user.email }}">
                {% endfor %}
            </datalist>

            <label for="message">Message :</label>
            <textarea name="message" rows="3" required></textarea>
            <button type="submit">Envoyer</button>
        </form>

        <button onclick="refreshMessages()">Rafraîchir</button>
        <button onclick="window.location.href='/messages/history'">Historique</button>

        <hr>
        <div id="messagesContainer">
            {% for msg in inbox %}
                <div class="message {% if msg.sender == current_user %}sent{% else %}received{% endif %}">
                    <div>{{ msg.message }}</div>
                    <div class="timestamp">{{ msg.timestamp.strftime('%d/%m/%Y %H:%M') if msg.timestamp else '' }}</div>
                    <div class="message-actions">
                        <button onclick="deleteMessage('{{ msg._id }}')">Supprimer</button>
                    </div>
                </div>
            {% endfor %}
        </div>
    </div>

    <script>
        function refreshMessages() {
            window.location.reload();
        }

        function deleteMessage(messageId) {
            fetch('/delete_message/' + messageId, { method: 'DELETE' })
                .then(response => response.json())
                .then(data => {
                    if (data.success) {
                        alert('Message supprimé avec succès!');
                        window.location.reload();
                    } else {
                        alert('Erreur lors de la suppression du message.');
                    }
                });
        }

        let previousMessageCount = {{ inbox|length }};

        function checkForNewMessages() {
            fetch("/messages/count")
                .then(response => response.json())
                .then(data => {
                    if (data.count > previousMessageCount) {
                        notifyNewMessage();
                    }
                })
                .catch(error => console.error('Error fetching messages:', error));
        }

        function notifyNewMessage() {
            const audio = new Audio("https://notificationsounds.com/storage/sounds/file-sounds-1153-pristine.mp3");
            audio.play();
            alert("📨 Nouveau message reçu !");
            location.reload();
        }

        setInterval(checkForNewMessages, 10000);
    </script>
</body>
</html>
"""

MESSAGE_HISTORY_TEMPLATE = """
<!DOCTYPE html>
<html lang="fr">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Historique des Messages</title>
    <style>
        :root {
            --primary-color: #8052e6;
            --primary-dark: #6a40d0;
            --secondary-color: #2c2c6c;
            --accent-color: #ff6b6b;
            --success-color: #4ecdc4;
            --warning-color: #ffe66d;
            --background-color: #f8fafc;
            --card-background: #ffffff;
            --text-primary: #2d3748;
            --text-secondary: #718096;
            --border-color: #e2e8f0;
            --shadow-light: 0 4px 6px rgba(0, 0, 0, 0.07);
            --shadow-medium: 0 10px 25px rgba(0, 0, 0, 0.1);
            --shadow-heavy: 0 20px 40px rgba(0, 0, 0, 0.15);
            --border-radius: 12px;
            --transition: all 0.3s cubic-bezier(0.4, 0, 0.2, 1);
        }
        * {
            margin: 0;
            padding: 0;
            box-sizing: border-box;
        }
        body {
            font-family: 'Segoe UI', sans-serif;
            background: var(--background-color);
            color: var(--text-primary);
            line-height: 1.6;
            padding: 2rem;
        }
        .history-container {
            max-width: 800px;
            margin: 0 auto;
            background: var(--card-background);
            padding: 2rem;
            border-radius: var(--border-radius);
            box-shadow: var(--shadow-medium);
            border: 1px solid var(--border-color);
        }
        h1 {
            text-align: center;
            color: var(--primary-color);
            margin-bottom: 2rem;
        }
        .message {
            padding: 1.5rem;
            margin-bottom: 1rem;
            border-radius: var(--border-radius);
            transition: var(--transition);
            position: relative;
            overflow: hidden;
        }
        .sent {
            background-color: var(--success-color);
            text-align: right;
            margin-left: 2rem;
        }
        .received {
            background-color: var(--warning-color);
            text-align: left;
            margin-right: 2rem;
        }
        .timestamp {
            font-size: 0.8rem;
            color: var(--text-secondary);
            margin-top: 0.5rem;
        }
        .back-button {
            display: block;
            width: 100%;
            padding: 1rem;
            background: linear-gradient(135deg, var(--primary-color), var(--primary-dark));
            color: white;
            border: none;
            border-radius: var(--border-radius);
            font-size: 1rem;
            font-weight: 600;
            cursor: pointer;
            transition: var(--transition);
            text-align: center;
            margin-top: 2rem;
        }
        .back-button:hover {
            transform: translateY(-2px);
            box-shadow: var(--shadow-medium);
        }
    </style>
</head>
<body>
    <div class="history-container">
        <h1>📜 Historique des Messages</h1>
        {% for msg in message_history %}
            <div class="message {% if msg.sender == current_user %}sent{% else %}received{% endif %}">
                <div><strong>De:</strong> {{ msg.sender }}</div>
                <div><strong>À:</strong> {{ msg.recipient }}</div>
                <div>{{ msg.message }}</div>
                <div class="timestamp">{{ msg.timestamp.strftime('%d/%m/%Y %H:%M') if msg.timestamp else '' }}</div>
            </div>
        {% endfor %}
        <button class="back-button" onclick="window.location.href='/messages'">Retour à la messagerie</button>
    </div>
</body>
</html>
"""


ANALYTICS_TEMPLATE = """
<!DOCTYPE html>
<html lang="fr">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Statistiques</title>
    <script src="https://cdn.jsdelivr.net/npm/chart.js"></script>
    <style>
        :root {
            --primary-color: #8052e6;
            --primary-dark: #6a40d0;
            --secondary-color: #2c2c6c;
            --accent-color: #ff6b6b;
            --success-color: #4ecdc4;
            --warning-color: #ffe66d;
            --background-color: #f8fafc;
            --card-background: #ffffff;
            --text-primary: #2d3748;
            --text-secondary: #718096;
            --border-color: #e2e8f0;
            --shadow-light: 0 4px 6px rgba(0, 0, 0, 0.07);
            --shadow-medium: 0 10px 25px rgba(0, 0, 0, 0.1);
            --shadow-heavy: 0 20px 40px rgba(0, 0, 0, 0.15);
            --border-radius: 12px;
            --transition: all 0.3s cubic-bezier(0.4, 0, 0.2, 1);
        }
        * {
            margin: 0;
            padding: 0;
            box-sizing: border-box;
        }
        body {
            font-family: 'Segoe UI', sans-serif;
            background: var(--background-color);
            color: var(--text-primary);
            line-height: 1.6;
            padding: 2rem;
        }
        .analytics-container {
            max-width: 1200px;
            margin: 0 auto;
            background: var(--card-background);
            padding: 2rem;
            border-radius: var(--border-radius);
            box-shadow: var(--shadow-medium);
            border: 1px solid var(--border-color);
        }
        h1 {
            text-align: center;
            color: var(--primary-color);
            margin-bottom: 2rem;
        }
        .chart-container {
            margin: 2rem 0;
        }
        .chart-container canvas {
            width: 100%;
            height: auto;
        }
        .download-button {
            background: linear-gradient(135deg, var(--primary-color), var(--primary-dark));
            color: white;
            border: none;
            padding: 1rem;
            border-radius: var(--border-radius);
            font-size: 1rem;
            font-weight: 600;
            cursor: pointer;
            transition: var(--transition);
            display: block;
            width: 100%;
            margin-top: 1rem;
            text-align: center;
        }
        .download-button:hover {
            transform: translateY(-2px);
            box-shadow: var(--shadow-medium);
        }
    </style>
</head>
<body>
    <div class="analytics-container">
        <h1>📊 Statistiques</h1>
        <div class="chart-container">
            <canvas id="statsChart"></canvas>
        </div>
        <div class="chart-container">
            <canvas id="pieChart"></canvas>
        </div>
        <button class="download-button" onclick="downloadPDF()">Télécharger en PDF</button>
    </div>

    <script>
        const ctx = document.getElementById('statsChart').getContext('2d');
        const statsChart = new Chart(ctx, {
            type: 'bar',
            data: {
                labels: ['Utilisateurs', 'Offres', 'Candidatures'],
                datasets: [{
                    label: 'Statistiques Globales',
                    data: [{{ total_users }}, {{ total_offers }}, {{ total_applications }}],
                    backgroundColor: [
                        'rgba(128, 82, 230, 0.7)',
                        'rgba(78, 205, 196, 0.7)',
                        'rgba(255, 107, 107, 0.7)'
                    ],
                    borderColor: [
                        'rgba(128, 82, 230, 1)',
                        'rgba(78, 205, 196, 1)',
                        'rgba(255, 107, 107, 1)'
                    ],
                    borderWidth: 1
                }]
            },
            options: {
                scales: {
                    y: {
                        beginAtZero: true
                    }
                }
            }
        });

        const pieCtx = document.getElementById('pieChart').getContext('2d');
        const pieChart = new Chart(pieCtx, {
            type: 'pie',
            data: {
                labels: ['Freelancers', 'Recruteurs', 'Administrateurs'],
                datasets: [{
                    label: 'Distribution des Utilisateurs',
                    data: [{{ freelancers_count }}, {{ recruiters_count }}, {{ admins_count }}],
                    backgroundColor: [
                        'rgba(128, 82, 230, 0.7)',
                        'rgba(78, 205, 196, 0.7)',
                        'rgba(255, 107, 107, 0.7)'
                    ],
                    borderColor: [
                        'rgba(128, 82, 230, 1)',
                        'rgba(78, 205, 196, 1)',
                        'rgba(255, 107, 107, 1)'
                    ],
                    borderWidth: 1
                }]
            },
            options: {
                responsive: true,
                plugins: {
                    legend: {
                        position: 'top',
                    },
                    title: {
                        display: true,
                        text: 'Distribution des Types d\'Utilisateurs'
                    }
                }
            }
        });

        function downloadPDF() {
            fetch('/download_stats_pdf')
                .then(response => response.blob())
                .then(blob => {
                    const url = window.URL.createObjectURL(blob);
                    const a = document.createElement('a');
                    a.href = url;
                    a.download = 'statistiques.pdf';
                    document.body.appendChild(a);
                    a.click();
                    document.body.removeChild(a);
                });
        }
    </script>
</body>
</html>
"""

SETTINGS_TEMPLATE = """
<!DOCTYPE html>
<html lang="fr">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Paramètres du Compte</title>
    <style>
        :root {
            --primary-color: #8052e6;
            --primary-dark: #6a40d0;
            --secondary-color: #2c2c6c;
            --accent-color: #ff6b6b;
            --success-color: #4ecdc4;
            --warning-color: #ffe66d;
            --background-color: #f8fafc;
            --card-background: #ffffff;
            --text-primary: #2d3748;
            --text-secondary: #718096;
            --border-color: #e2e8f0;
            --shadow-light: 0 4px 6px rgba(0, 0, 0, 0.07);
            --shadow-medium: 0 10px 25px rgba(0, 0, 0, 0.1);
            --shadow-heavy: 0 20px 40px rgba(0, 0, 0, 0.15);
            --border-radius: 12px;
            --transition: all 0.3s cubic-bezier(0.4, 0, 0.2, 1);
        }
        * {
            margin: 0;
            padding: 0;
            box-sizing: border-box;
        }
        body {
            font-family: 'Segoe UI', sans-serif;
            background: var(--background-color);
            color: var(--text-primary);
            line-height: 1.6;
            padding: 2rem;
        }
        .settings-container {
            max-width: 600px;
            margin: 0 auto;
            background: var(--card-background);
            padding: 2rem;
            border-radius: var(--border-radius);
            box-shadow: var(--shadow-medium);
            border: 1px solid var(--border-color);
        }
        h1 {
            text-align: center;
            color: var(--primary-color);
            margin-bottom: 2rem;
        }
        .form-group {
            margin-bottom: 1.5rem;
        }
        .form-group label {
            display: block;
            margin-bottom: 0.5rem;
            font-weight: 600;
            color: var(--text-primary);
        }
        .form-group input {
            width: 100%;
            padding: 1rem;
            border: 2px solid var(--border-color);
            border-radius: var(--border-radius);
            font-size: 1rem;
            transition: var(--transition);
        }
        .form-group input:focus {
            outline: none;
            border-color: var(--primary-color);
            box-shadow: 0 0 0 3px rgba(128, 82, 230, 0.1);
        }
        .submit-button {
            background: linear-gradient(135deg, var(--primary-color), var(--primary-dark));
            color: white;
            border: none;
            padding: 1rem;
            border-radius: var(--border-radius);
            font-size: 1rem;
            font-weight: 600;
            cursor: pointer;
            transition: var(--transition);
            display: block;
            width: 100%;
            margin-top: 1rem;
        }
        .submit-button:hover {
            transform: translateY(-2px);
            box-shadow: var(--shadow-medium);
        }
    </style>
</head>
<body>
    <div class="settings-container">
        <h1>⚙️ Paramètres du Compte</h1>
        <form id="settingsForm" method="POST">
            <div class="form-group">
                <label for="email">E-mail</label>
                <input type="email" id="email" name="email" value="{{ user.email }}" readonly>
            </div>
            <div class="form-group">
                <label for="full_name">Nom Complet</label>
                <input type="text" id="full_name" name="full_name" value="{{ user.full_name }}" required>
            </div>
            <div class="form-group">
                <label for="new_password">Nouveau Mot de Passe</label>
                <input type="password" id="new_password" name="new_password" placeholder="Entrez un nouveau mot de passe">
            </div>
            <button type="submit" class="submit-button">Mettre à jour</button>
        </form>
    </div>

    <script>
        document.getElementById('settingsForm').addEventListener('submit', function(event) {
            let isValid = true;
            const inputs = document.querySelectorAll('input[required]');

            inputs.forEach(input => {
                if (!input.value.trim()) {
                    isValid = false;
                    input.style.borderColor = 'var(--accent-color)';
                } else {
                    input.style.borderColor = 'var(--border-color)';
                }
            });

            if (!isValid) {
                event.preventDefault();
                alert('Veuillez remplir tous les champs obligatoires.');
            }
        });
    </script>
</body>
</html>
"""

RECRUITER_SETTINGS_TEMPLATE = """
<!DOCTYPE html>
<html lang="fr">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Paramètres du Compte</title>
    <style>
        :root {
            --primary-color: #8052e6;
            --primary-dark: #6a40d0;
            --secondary-color: #2c2c6c;
            --accent-color: #ff6b6b;
            --success-color: #4ecdc4;
            --warning-color: #ffe66d;
            --background-color: #f8fafc;
            --card-background: #ffffff;
            --text-primary: #2d3748;
            --text-secondary: #718096;
            --border-color: #e2e8f0;
            --shadow-light: 0 4px 6px rgba(0, 0, 0, 0.07);
            --shadow-medium: 0 10px 25px rgba(0, 0, 0, 0.1);
            --shadow-heavy: 0 20px 40px rgba(0, 0, 0, 0.15);
            --border-radius: 12px;
            --transition: all 0.3s cubic-bezier(0.4, 0, 0.2, 1);
        }
        * {
            margin: 0;
            padding: 0;
            box-sizing: border-box;
        }
        body {
            font-family: 'Segoe UI', sans-serif;
            background: var(--background-color);
            color: var(--text-primary);
            line-height: 1.6;
            padding: 2rem;
        }
        .settings-container {
            max-width: 600px;
            margin: 0 auto;
            background: var(--card-background);
            padding: 2rem;
            border-radius: var(--border-radius);
            box-shadow: var(--shadow-medium);
            border: 1px solid var(--border-color);
        }
        h1 {
            text-align: center;
            color: var(--primary-color);
            margin-bottom: 2rem;
        }
        .form-group {
            margin-bottom: 1.5rem;
        }
        .form-group label {
            display: block;
            margin-bottom: 0.5rem;
            font-weight: 600;
            color: var(--text-primary);
        }
        .form-group input, .form-group textarea {
            width: 100%;
            padding: 1rem;
            border: 2px solid var(--border-color);
            border-radius: var(--border-radius);
            font-size: 1rem;
            transition: var(--transition);
        }
        .form-group input:focus, .form-group textarea:focus {
            outline: none;
            border-color: var(--primary-color);
            box-shadow: 0 0 0 3px rgba(128, 82, 230, 0.1);
        }
        .form-group textarea {
            min-height: 150px;
            resize: vertical;
        }
        .submit-button {
            background: linear-gradient(135deg, var(--primary-color), var(--primary-dark));
            color: white;
            border: none;
            padding: 1rem;
            border-radius: var(--border-radius);
            font-size: 1rem;
            font-weight: 600;
            cursor: pointer;
            transition: var(--transition);
            display: block;
            width: 100%;
            margin-top: 1rem;
        }
        .submit-button:hover {
            transform: translateY(-2px);
            box-shadow: var(--shadow-medium);
        }
    </style>
</head>
<body>
    <div class="settings-container">
        <h1>⚙️ Paramètres du Compte</h1>
        <form id="settingsForm" method="POST">
            <div class="form-group">
                <label for="nom">Nom</label>
                <input type="text" id="nom" name="nom" value="{{ recruteur.nom }}" required>
            </div>
            <div class="form-group">
                <label for="description">Description</label>
                <textarea id="description" name="description" required>{{ recruteur.description }}</textarea>
            </div>
            <div class="form-group">
                <label for="password">Mot de Passe</label>
                <input type="password" id="password" name="password" placeholder="Laisser vide pour ne pas changer">
            </div>
            <button type="submit" class="submit-button">Mettre à jour</button>
        </form>
    </div>

    <script>
        document.getElementById('settingsForm').addEventListener('submit', function(event) {
            let isValid = true;
            const inputs = document.querySelectorAll('input[required], textarea[required]');

            inputs.forEach(input => {
                if (!input.value.trim()) {
                    isValid = false;
                    input.style.borderColor = 'var(--accent-color)';
                } else {
                    input.style.borderColor = 'var(--border-color)';
                }
            });

            if (!isValid) {
                event.preventDefault();
                alert('Veuillez remplir tous les champs obligatoires.');
            }
        });
    </script>
</body>
</html>
"""
CANDIDATS_PROJETS_TEMPLATE = """
<!DOCTYPE html>
<html lang="fr">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Candidats et Projets</title>
    <style>
        :root {
            --primary-color: #6a40d0;
            --secondary-color: #f8f9fa;
            --accent-color: #007BFF;
            --success-color: #28a745;
            --danger-color: #dc3545;
            --dark-color: #343a40;
            --light-color: #f8f9fa;
            --text-color: #495057;
            --card-shadow: 0 4px 20px rgba(0, 0, 0, 0.1);
            --transition: all 0.3s ease;
        }

        body {
            font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
            padding: 2rem;
            background-color: var(--secondary-color);
            color: var(--text-color);
            line-height: 1.6;
        }

        h1 {
            color: var(--dark-color);
            margin-bottom: 1.5rem;
            font-weight: 600;
        }

        .search-box {
            margin-bottom: 2rem;
        }

        .search-box input {
            padding: 12px 20px;
            width: 100%;
            max-width: 400px;
            font-size: 16px;
            border: 2px solid #dee2e6;
            border-radius: 30px;
            transition: var(--transition);
            box-shadow: 0 4px 6px rgba(0, 0, 0, 0.1);
        }

        .search-box input:focus {
            outline: none;
            border-color: var(--primary-color);
            box-shadow: 0 0 0 3px rgba(106, 64, 208, 0.2);
        }

        .card {
            background: white;
            border-radius: 10px;
            padding: 1.5rem;
            margin-bottom: 1.5rem;
            box-shadow: var(--card-shadow);
            transition: var(--transition);
            border-left: 4px solid var(--primary-color);
        }

        .card:hover {
            transform: translateY(-5px);
            box-shadow: 0 10px 25px rgba(0, 0, 0, 0.15);
        }

        .section {
            margin-top: 1rem;
        }

        .section h3 {
            color: var(--primary-color);
            margin-bottom: 0.75rem;
        }

        .cv-link {
            margin-top: 0.5rem;
            display: inline-block;
            color: var(--accent-color);
            text-decoration: none;
            font-weight: 500;
        }

        .cv-link:hover {
            text-decoration: underline;
        }

        .tag {
            background: #e9ecef;
            padding: 4px 10px;
            border-radius: 20px;
            margin-right: 5px;
            font-size: 0.85em;
            color: var(--dark-color);
        }

        .action-buttons {
            margin-top: 1.5rem;
            display: flex;
            gap: 10px;
        }

        .action-buttons button {
            padding: 10px 20px;
            border: none;
            border-radius: 20px;
            font-weight: bold;
            cursor: pointer;
            transition: var(--transition);
            box-shadow: 0 2px 5px rgba(0, 0, 0, 0.1);
        }

        .btn-validate {
            background-color: var(--success-color);
            color: white;
        }

        .btn-validate:hover {
            background-color: #218838;
            transform: translateY(-2px);
        }

        .btn-reject {
            background-color: var(--danger-color);
            color: white;
        }

        .btn-reject:hover {
            background-color: #c82333;
            transform: translateY(-2px);
        }

        @media (max-width: 768px) {
            body {
                padding: 1rem;
            }

            .action-buttons {
                flex-direction: column;
            }

            .action-buttons button {
                width: 100%;
                margin-bottom: 10px;
            }
        }
    </style>
</head>
<body>
    <h1>👥 Liste des candidats & projets</h1>
    <div class="search-box">
        <input type="text" id="searchInput" placeholder="🔍 Rechercher un candidat...">
    </div>
    <div id="candidatesContainer">
        {% for email, infos in data.items() %}
            <div class="card candidate-card">
                <h2>{{ infos.profile.full_name }} <small style="color: var(--dark-color); opacity: 0.7;">{{ email }}</small></h2>
                <div>
                    <strong>Statut actuel :</strong>
                    {% if infos.status == "Validée" %}
                        <span class="tag" style="background-color: var(--success-color); color: white;">✅ Validée</span>
                    {% elif infos.status == "Refusée" %}
                        <span class="tag" style="background-color: var(--danger-color); color: white;">❌ Refusée</span>
                    {% else %}
                        <span class="tag" style="background-color: #ffc107; color: white;">🕐 En cours</span>
                    {% endif %}
                </div>
                <div class="section">
                    <h3>📄 CV</h3>
                    {% if infos.cv and 'url' in infos.cv %}
                        <a href="{{ infos.cv.url }}" class="cv-link" target="_blank">📎 Voir le CV</a>
                    {% else %}
                        <p>Aucun CV disponible</p>
                    {% endif %}
                </div>
                <div class="section">
                    <h3>🧠 Résultat Matching</h3>
                    {% if infos.matching %}
                        <pre style="background: #f8f9fa; padding: 1rem; border-radius: 5px; overflow-x: auto;">{{ infos.matching.match_result }}</pre>
                    {% else %}
                        <p>Pas encore de résultat de matching</p>
                    {% endif %}
                </div>
                <div class="section">
                    <h3>🚀 Projets</h3>
                    {% if infos.projects %}
                        <ul style="list-style-type: none; padding: 0;">
                            {% for p in infos.projects %}
                                <li style="background: #f8f9fa; padding: 1rem; margin-bottom: 0.5rem; border-radius: 5px;">
                                    <strong>{{ p.title }}</strong> — {{ p.description }}<br>
                                    <small>🔗 <a href="{{ p.url }}" target="_blank" style="color: var(--accent-color);">{{ p.url }}</a></small>
                                </li>
                            {% endfor %}
                        </ul>
                    {% else %}
                        <p>Aucun projet ajouté</p>
                    {% endif %}
                </div>
                <div class="action-buttons">
                    <button class="btn-validate" onclick="updateStatus('{{ email }}', 'Validée')">✅ Valider</button>
                    <button class="btn-reject" onclick="updateStatus('{{ email }}', 'Refusée')">❌ Refuser</button>
                </div>
            </div>
        {% endfor %}
    </div>
    <script>
        function updateStatus(email, status) {
            fetch('/update_candidate_status', {
                method: 'POST',
                headers: { 'Content-Type': 'application/json' },
                body: JSON.stringify({ email: email, status: status })
            })
            .then(res => res.json())
            .then(data => {
                alert(data.message);
                location.reload();
            });
        }

        // Recherche dynamique
        const input = document.getElementById("searchInput");
        input.addEventListener("keyup", function () {
            const filter = input.value.toLowerCase();
            const cards = document.getElementsByClassName("candidate-card");
            Array.from(cards).forEach(card => {
                const name = card.querySelector("h2").innerText.toLowerCase();
                card.style.display = name.includes(filter) ? "block" : "none";
            });
        });
    </script>
</body>
</html>

"""



@app.route('/')
def homepage():
    return render_template_string(HOMEPAGE_TEMPLATE)

@app.route('/login')
def login():
    return render_template_string(LOGIN_TEMPLATE)

@app.route('/register')
def register():
    return render_template_string(REGISTER_TEMPLATE)

@app.route('/authenticate', methods=['POST'])
def authenticate():
    email = request.form.get('email')
    password = request.form.get('password')

    user = users_collection.find_one({"email": email})

    if user and check_password_hash(user['password'], password):
        session['user_id'] = str(user['_id'])
        session['user_type'] = user['user_type']
        session['user_email'] = user['email']

        # Ajout de notifications
        if user['user_type'] == "recruiter":
            add_notification("admin", f"👤 Recruteur {user['email']} s’est connecté", None)
        elif user['user_type'] == "freelance":
            add_notification("recruiter", f"👤 Étudiant {user['email']} s’est connecté", None)

        return redirect(url_for('dashboard', user_type=user['user_type']))
    else:
        return redirect(url_for('login'))

@app.route('/register', methods=['POST'])
def register_user():
    email = request.form.get('email')
    full_name = request.form.get('full_name')
    user_type = request.form.get('user_type')
    password = request.form.get('password')

    existing_user = users_collection.find_one({"email": email})
    if existing_user:
        return "Email already registered", 400

    hashed_password = generate_password_hash(password)

    new_user = {
        "email": email,
        "full_name": full_name,
        "user_type": user_type,
        "password": hashed_password
    }
    users_collection.insert_one(new_user)

    return redirect(url_for('login'))

@app.route('/dashboard/<user_type>')
def dashboard(user_type):
    if 'user_type' not in session:
        return redirect(url_for('login'))

    if user_type == "admin":
        total_users = users_collection.count_documents({})
        total_offers = offers_collection.count_documents({})
        total_applications = applications_collection.count_documents({})
        users = list(users_collection.find({}))
        for user in users:
            user["_id"] = str(user["_id"])
        return render_template_string(ADMIN_DASHBOARD_TEMPLATE, total_users=total_users, total_offers=total_offers, total_applications=total_applications, users=users)
    elif user_type == "freelance":
        return render_template_string(STUDENT_DASHBOARD_TEMPLATE)
    elif user_type == "recruiter":
        return render_template_string(RECRUITER_DASHBOARD_TEMPLATE)
    return f"Bienvenue sur le tableau de bord pour {user_type}!"

@app.route('/matching_offers/<skill>')
def matching_offers(skill):
    matching = list(offers_collection.find({"matching_tags": {"$in": [skill.lower()]}}))
    for o in matching:
        o["_id"] = str(o["_id"])
    return {"matches": matching}

@app.route('/api/offers')
def api_offers():
    offers = list(offers_collection.find({"status": "active"}))
    for offer in offers:
        offer["_id"] = str(offer["_id"])
    return {"offers": offers}

@app.route('/create_offer', methods=['GET', 'POST'])
def create_offer():
    if 'user_type' not in session or session['user_type'] != 'recruiter':
        return redirect(url_for('login'))

    if request.method == 'POST':
        title = request.form.get('title')
        description = request.form.get('description')
        matching_tags = request.form.get('matching_tags', '').split(',')

        offer = {
            "title": title,
            "description": description,
            "status": "active",
            "created_by": session.get('user_email', 'recruteur inconnu'),
            "matching_tags": [tag.strip().lower() for tag in matching_tags],
            "createdAt": datetime.now(),
            "recruiter_id": session.get('user_id')
        }
        offers_collection.insert_one(offer)

        # Ajout de notifications
        add_notification("freelance", "🆕 Nouvelle offre publiée par un recruteur")

        return redirect(url_for('dashboard', user_type='recruiter'))

    return render_template_string("""
<!DOCTYPE html>
<html lang="fr">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Nouvelle Offre</title>
    <style>
        :root {
            --primary-color: #8052e6;
            --primary-dark: #6a40d0;
            --secondary-color: #2c2c6c;
            --accent-color: #ff6b6b;
            --success-color: #4ecdc4;
            --warning-color: #ffe66d;
            --background-color: #f8fafc;
            --card-background: #ffffff;
            --text-primary: #2d3748;
            --text-secondary: #718096;
            --border-color: #e2e8f0;
            --shadow-light: 0 4px 6px rgba(0, 0, 0, 0.07);
            --shadow-medium: 0 10px 25px rgba(0, 0, 0, 0.1);
            --shadow-heavy: 0 20px 40px rgba(0, 0, 0, 0.15);
            --border-radius: 12px;
            --transition: all 0.3s cubic-bezier(0.4, 0, 0.2, 1);
        }
        * {
            margin: 0;
            padding: 0;
            box-sizing: border-box;
        }
        body {
            font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
            background: var(--background-color);
            color: var(--text-primary);
            line-height: 1.6;
        }
        .form-container {
            background: var(--card-background);
            padding: 2rem;
            border-radius: var(--border-radius);
            box-shadow: var(--shadow-light);
            border: 1px solid var(--border-color);
            transition: var(--transition);
            max-width: 600px;
            margin: 2rem auto;
        }
        .form-container h2 {
            color: var(--primary-color);
            margin-bottom: 1.5rem;
            text-align: center;
        }
        .form-group {
            margin-bottom: 1.5rem;
        }
        .form-group label {
            display: block;
            margin-bottom: 0.5rem;
            font-weight: 600;
            color: var(--text-primary);
        }
        .form-group input,
        .form-group textarea {
            width: 100%;
            padding: 1rem;
            border: 2px solid var(--border-color);
            border-radius: var(--border-radius);
            font-size: 1rem;
            transition: var(--transition);
        }
        .form-group input:focus,
        .form-group textarea:focus {
            outline: none;
            border-color: var(--primary-color);
            box-shadow: 0 0 0 3px rgba(128, 82, 230, 0.1);
        }
        .form-group textarea {
            min-height: 150px;
            resize: vertical;
        }
        .submit-button {
            background: linear-gradient(135deg, var(--primary-color), var(--primary-dark));
            color: white;
            border: none;
            padding: 1rem;
            border-radius: var(--border-radius);
            font-size: 1rem;
            font-weight: 600;
            cursor: pointer;
            transition: var(--transition);
            display: block;
            width: 100%;
            margin-top: 1rem;
        }
        .submit-button:hover {
            transform: translateY(-2px);
            box-shadow: var(--shadow-medium);
        }
    </style>
</head>
<body>
    <div class="form-container">
        <h2>Nouvelle Offre</h2>
        <form method="post">
            <div class="form-group">
                <label for="title">Titre:</label>
                <input type="text" id="title" name="title" required>
            </div>
            <div class="form-group">
                <label for="description">Description:</label>
                <textarea id="description" name="description" required></textarea>
            </div>
            <div class="form-group">
                <label for="matching_tags">Tags (ex: python,react,data):</label>
                <input type="text" id="matching_tags" name="matching_tags">
            </div>
            <button type="submit" class="submit-button">Publier</button>
        </form>
    </div>
</body>
</html>
""")

@app.route('/complete_profile', methods=['GET', 'POST'])
def complete_profile():
    if 'user_type' not in session or session['user_type'] != 'recruiter':
        return redirect(url_for('login'))

    if request.method == 'POST':
        data = {
            "email": session.get("user_email"),
            "company_name": request.form['company_name'],
            "sector": request.form['sector'],
            "size": request.form['size'],
            "description": request.form['description'],
            "logo_url": request.form['logo_url']
        }
        db.company_profiles.update_one({"email": data["email"]}, {"$set": data}, upsert=True)
        return redirect(url_for('dashboard', user_type='recruiter'))

    return render_template_string(COMPLETE_PROFILE_TEMPLATE)

@app.route('/logout')
def logout():
    session.pop('user_id', None)
    session.pop('user_type', None)
    session.pop('user_email', None)
    return redirect(url_for('homepage'))

@app.route('/messages', methods=['GET', 'POST'])
def messages():
    if 'user_type' not in session:
        return redirect(url_for('login'))

    user_email = session.get('user_email', '')
    selected_user = request.args.get('user')

    if request.method == 'POST':
        recipient = request.form.get('recipient')
        message = request.form.get('message')
        messages_collection.insert_one({
            "sender": user_email,
            "recipient": recipient,
            "message": message,
            "timestamp": datetime.now()
        })
        return redirect(url_for('messages', user=recipient))

    if selected_user:
        inbox = list(messages_collection.find({
            "$or": [
                {"sender": user_email, "recipient": selected_user},
                {"sender": selected_user, "recipient": user_email}
            ]
        }).sort("timestamp", 1))
    else:
        inbox = list(messages_collection.find({
            "$or": [
                {"sender": user_email},
                {"recipient": user_email}
            ]
        }).sort("timestamp", -1))

    all_users = list(users_collection.find({}, {"email": 1}))

    return render_template_string(MESSAGE_TEMPLATE, inbox=inbox, current_user=user_email, all_users=all_users, selected_user=selected_user)

@app.route('/messages/history')
def message_history():
    if 'user_type' not in session:
        return redirect(url_for('login'))

    user_email = session.get('user_email', '')
    message_history = list(messages_collection.find({
        "$or": [
            {"sender": user_email},
            {"recipient": user_email}
        ]
    }).sort("timestamp", -1))

    return render_template_string(MESSAGE_HISTORY_TEMPLATE, message_history=message_history, current_user=user_email)

@app.route('/messages/count')
def message_count():
    user_email = session.get('user_email', '')
    count = messages_collection.count_documents({
        "$or": [
            {"sender": user_email},
            {"recipient": user_email}
        ]
    })
    return {"count": count}

@app.route('/delete_message/<message_id>', methods=['DELETE'])
def delete_message(message_id):
    messages_collection.delete_one({"_id": ObjectId(message_id)})
    return jsonify({"success": True})

@app.route('/edit_offer/<offer_id>', methods=['GET', 'POST'])
def edit_offer(offer_id):
    if 'user_type' not in session or session['user_type'] != 'recruiter':
        return redirect(url_for('login'))

    offer = offers_collection.find_one({"_id": ObjectId(offer_id)})
    if not offer:
        return "Offre introuvable", 404

    if request.method == 'POST':
        title = request.form.get('title')
        description = request.form.get('description')
        matching_tags = request.form.get('matching_tags', '').split(',')

        offers_collection.update_one({"_id": ObjectId(offer_id)}, {"$set": {
            "title": title,
            "description": description,
            "matching_tags": [tag.strip().lower() for tag in matching_tags]
        }})

        return render_template_string("""
<html>
<head>
    <meta http-equiv="refresh" content="2; url=/dashboard/recruiter" />
    <style>
        body {
            background: #f8fafc;
            font-family: 'Segoe UI', sans-serif;
            text-align: center;
            padding: 3rem;
        }
        .message {
            background: #4ecdc4;
            color: white;
            padding: 1rem 2rem;
            border-radius: 10px;
            display: inline-block;
            font-size: 1.2rem;
            box-shadow: 0 4px 10px rgba(0,0,0,0.1);
        }
    </style>
</head>
<body>
    <div class="message">✅ Offre mise à jour avec succès !<br>Redirection en cours...</div>
</body>
</html>
""")

    return render_template_string(f"""
<!DOCTYPE html>
<html lang="fr">
<head>
    <meta charset="UTF-8">
    <title>Modifier l’Offre</title>
    <style>
        body {{
            background: #f8fafc;
            font-family: 'Segoe UI', sans-serif;
        }}
        .form-container {{
            background: #ffffff;
            max-width: 600px;
            margin: 3rem auto;
            padding: 2rem;
            border-radius: 12px;
            box-shadow: 0 10px 25px rgba(0,0,0,0.08);
        }}
        h2 {{
            color: #8052e6;
            margin-bottom: 1.5rem;
            text-align: center;
        }}
        .form-group {{
            margin-bottom: 1.2rem;
        }}
        label {{
            font-weight: bold;
            display: block;
            margin-bottom: 0.5rem;
            color: #333;
        }}
        input, textarea {{
            width: 100%;
            padding: 0.9rem;
            border: 2px solid #e2e8f0;
            border-radius: 8px;
            font-size: 1rem;
        }}
        input:focus, textarea:focus {{
            outline: none;
            border-color: #8052e6;
            box-shadow: 0 0 0 3px rgba(128, 82, 230, 0.1);
        }}
        .submit-button {{
            background: linear-gradient(135deg, #8052e6, #6a40d0);
            color: white;
            border: none;
            padding: 1rem;
            border-radius: 8px;
            font-size: 1rem;
            font-weight: bold;
            cursor: pointer;
            transition: all 0.3s ease;
            display: block;
            width: 100%;
            margin-top: 1rem;
        }}
        .submit-button:hover {{
            transform: translateY(-2px);
            box-shadow: 0 8px 25px rgba(0, 0, 0, 0.15);
        }}
    </style>
</head>
<body>
    <div class="form-container">
        <h2>✏️ Modifier l’offre</h2>
        <form method="post">
            <div class="form-group">
                <label for="title">Titre de l'offre :</label>
                <input type="text" id="title" name="title" value="{offer['title']}" required>
            </div>
            <div class="form-group">
                <label for="description">Description :</label>
                <textarea id="description" name="description" required>{offer['description']}</textarea>
            </div>
            <div class="form-group">
                <label for="matching_tags">Tags (ex : python,react,data):</label>
                <input type="text" id="matching_tags" name="matching_tags" value="{','.join(offer['matching_tags'])}">
            </div>
            <button type="submit" class="submit-button">✅ Enregistrer les modifications</button>
        </form>
    </div>
</body>
</html>
""")

@app.route('/delete_offer/<offer_id>')
def delete_offer(offer_id):
    offers_collection.delete_one({"_id": ObjectId(offer_id)})
    return redirect(url_for('dashboard', user_type='recruiter'))

@app.route('/upload_cv', methods=['POST'])
def upload_cv():
    if 'user_type' not in session or session['user_type'] != 'freelance':
        return jsonify({"error": "Unauthorized"}), 403

    if 'cv' not in request.files:
        return jsonify({"success": False, "message": "Aucun fichier reçu"})

    file = request.files['cv']
    if file.filename == '':
        return jsonify({"success": False, "message": "Aucun fichier sélectionné"})

    filename = file.filename
    ext = os.path.splitext(filename)[1]
    if ext not in ['.pdf', '.doc', '.docx']:
        return jsonify({"success": False, "message": "Format non supporté"})

    email = session.get("user_email")
    cvs_collection.insert_one({
        "email": email,
        "filename": filename,
        "content": "",
        "uploaded_at": datetime.utcnow(),
        "analyzed": False
    })

    return jsonify({"success": True, "filename": filename})

@app.route('/get_latest_cv')
def get_latest_cv():
    email = session.get("user_email")
    if not email:
        return jsonify({})
    cv = cvs_collection.find_one({"email": email}, sort=[("uploaded_at", -1)])
    return jsonify({"filename": cv['filename'] if cv else None})
from datetime import datetime, timezone

import re
from datetime import datetime, timezone

@app.route('/match_cv_to_offers', methods=['POST'])
def match_cv_to_offers():
    email = session.get("user_email")
    if not email:
        return jsonify({"error": "Utilisateur non connecté."})

    # 1. Récupérer le CV avec contenu réel
    cv_doc = cvs_collection.find_one({"email": email}, sort=[("uploaded_at", -1)])
    if not cv_doc or "content" not in cv_doc:
        return jsonify({"error": "Aucun CV ou contenu texte trouvé."})
    cv_text = cv_doc["content"]

    # 🔍 Tentative d'extraction du nom complet depuis le contenu du CV
    full_name = "Nom inconnu"
    civilite = "Candidat·e"

    # Cherche des indices de civilité + nom
    match = re.search(r'(Monsieur|Madame|Mademoiselle)\s+([A-Z][a-z]+)\s+([A-Z][a-z]+)', cv_text)
    if match:
        civilite = match.group(1)
        full_name = f"{match.group(2)} {match.group(3)}"
    else:
        # Sinon cherche une ligne avec "Nom :"
        match_alt = re.search(r'Nom\s*:\s*([A-Z][a-z]+)\s+([A-Z][a-z]+)', cv_text)
        if match_alt:
            full_name = f"{match_alt.group(1)} {match_alt.group(2)}"
            civilite = "Candidat·e"

    nom_complet_formaté = f"{civilite} {full_name}"

    # 2. Récupérer les projets de l'utilisateur
    user = users_collection.find_one({"email": email})
    if not user:
        return jsonify({"error": "Utilisateur introuvable."})
    user_id = str(user["_id"])

    projects = projects_collection.find({"user_id": user_id})
    projects_text = "\n".join([
        f"- {proj['title']} : {proj.get('description', '')} (Techs: {proj.get('technologies', '')})"
        for proj in projects
    ]) or "Aucun projet déclaré."

    # 3. Récupérer les offres réelles
    recent_offers = list(offers_collection.find().sort("createdAt", -1).limit(3))
    if not recent_offers:
        return jsonify({"error": "Aucune offre trouvée."})

    offers_text = "\n".join([
        f"🔹 {offer.get('title', 'Offre sans titre')} : {offer.get('description', '...')}"
        for offer in recent_offers
    ])
    offers_used_titles = [o.get("title", "Offre") for o in recent_offers]

    # 4. Construire le prompt avec le nom et la civilité imposés
    prompt = f"""
Tu dois analyser un profil dont le nom est **{nom_complet_formaté}**.

📄 Contenu du CV :
{cv_text}

📁 Projets :
{projects_text}

📦 Offres disponibles :
{offers_text}

Ta mission :
1. Commence ta réponse par : "{nom_complet_formaté},"
2. Pour chaque offre :
   - Dis si le matching est bon ou non
   - Donne un score sur 100% (pondération : Compétences 25%, Expérience 25%, Formation 25%, Projets 15%, Certifications 5%, Softskills 5%)
3. Résume tout cela en **5 lignes maximum**, de manière directe, claire, humaine et professionnelle.
🛑 Ne dépasse pas 5 lignes. Sois synthétique.
"""

    headers = {
        "Authorization": "Bearer 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",
        "Content-Type": "application/json"
    }

    payload = {
        "model": "meta-llama/llama-4-scout-17b-16e-instruct",
        "messages": [{"role": "user", "content": prompt}]
    }

    response = requests.post("https://api.groq.com/openai/v1/chat/completions", headers=headers, json=payload)
    result = response.json()
    match_result = result["choices"][0]["message"]["content"]

    # 5. Sauvegarde MongoDB
    matching_results_collection.insert_one({
        "email": email,
        "user_id": user_id,
        "result": match_result,
        "offers_used": offers_used_titles,
        "full_name": full_name,
        "civilite": civilite,
        "timestamp": datetime.now(timezone.utc)
    })

    return jsonify({
        "match_result": match_result,
        "offers_used": offers_used_titles
    })


@app.route('/ask_ai', methods=['POST'])
def ask_ai():
    q = request.json.get("question", "")
    headers = {
        "Authorization": "Bearer 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",
        "Content-Type": "application/json"
    }
    payload = {
        "model": "meta-llama/llama-4-scout-17b-16e-instruct",
        "messages": [{"role": "user", "content": q}]
    }
    response = requests.post(
        "https://api.groq.com/openai/v1/chat/completions",
        headers=headers,
        json=payload
    )
    result = response.json()
    return jsonify({"answer": result["choices"][0]["message"]["content"][:700]})

@app.route('/add_project', methods=['POST'])
def add_project():
    email = session.get("user_email")
    project = {
        "email": email,
        "title": request.form["title"],
        "description": request.form["description"],
        "technologies": request.form["technologies"].split(","),
        "url": request.form["url"],
        "created_at": datetime.utcnow()
    }
    projects_collection.insert_one(project)
    return jsonify({"status": "success"})

@app.route('/get_projects')
def get_projects():
    email = session.get("user_email")
    projects = list(projects_collection.find({"email": email}))
    for project in projects:
        project["_id"] = str(project["_id"])
    return jsonify(projects)

@app.route('/api/get_recent_offers')
def get_recent_offers():
    offres = offers_collection.find().sort("createdAt", -1).limit(5)
    return jsonify([
        {
            "title": o.get("title", "Sans titre"),
            "description": o.get("description", ""),
            "tags": o.get("matching_tags", [])
        } for o in offres
    ])

@app.route('/api/candidates')
def get_recent_candidates():
    recruiter_id = session.get('user_id')
    candidates = list(applications_collection.find({"recruiter_id": recruiter_id}).sort("appliedAt", -1).limit(5))
    for candidate in candidates:
        candidate["_id"] = str(candidate["_id"])
    return {"candidates": candidates}

@app.route('/api/projects')
def get_recent_projects():
    recruiter_id = session.get('user_id')
    projects = list(projects_collection.find({"recruiter_id": recruiter_id}).sort("created_at", -1).limit(5))
    for project in projects:
        project["_id"] = str(project["_id"])
    return {"projects": projects}

@app.route('/api/real_time_applications')
def get_real_time_applications():
    total_applications = applications_collection.count_documents({})
    return {"totalApplications": total_applications}

@app.route('/analytics')
def analytics():
    total_users = users_collection.count_documents({})
    total_offers = offers_collection.count_documents({})
    total_applications = applications_collection.count_documents({})
    freelancers_count = users_collection.count_documents({"user_type": "freelance"})
    recruiters_count = users_collection.count_documents({"user_type": "recruiter"})
    admins_count = users_collection.count_documents({"user_type": "admin"})

    return render_template_string(ANALYTICS_TEMPLATE, total_users=total_users, total_offers=total_offers, total_applications=total_applications, freelancers_count=freelancers_count, recruiters_count=recruiters_count, admins_count=admins_count)

@app.route('/download_stats_pdf')
def download_stats_pdf():
    buffer = io.BytesIO()
    doc = SimpleDocTemplate(buffer, pagesize=letter, title="Statistiques")

    styles = getSampleStyleSheet()
    elements = []

    # Title
    title = Paragraph("Statistiques Globales", styles['Title'])
    elements.append(title)
    elements.append(Spacer(1, 12))

    # Bar Chart
    data = [
        ["Catégorie", "Nombre"],
        ["Utilisateurs", users_collection.count_documents({})],
        ["Offres", offers_collection.count_documents({})],
        ["Candidatures", applications_collection.count_documents({})]
    ]

    drawing = Drawing(400, 200)
    bc = VerticalBarChart()
    bc.x = 50
    bc.y = 50
    bc.width = 300
    bc.height = 150
    bc.data = [list(row) for row in data[1:]]
    bc.categoryAxis.categoryNames = data[0][1:]
    bc.valueAxis.valueMin = 0
    bc.valueAxis.valueMax = max([row[1] for row in data[1:]]) + 10
    bc.valueAxis.valueStep = 10
    bc.valueAxis.labelTextFormat = '%d'
    drawing.add(bc)
    elements.append(drawing)

    # Pie Chart
    pie_data = [
        ["Freelancers", users_collection.count_documents({"user_type": "freelance"})],
        ["Recruteurs", users_collection.count_documents({"user_type": "recruiter"})],
        ["Administrateurs", users_collection.count_documents({"user_type": "admin"})]
    ]

    pie_drawing = Drawing(400, 200)
    pc = Pie()
    pc.x = 150
    pc.y = 50
    pc.width = 150
    pc.height = 150
    pc.data = [row[1] for row in pie_data]
    pc.labels = [row[0] for row in pie_data]
    pie_drawing.add(pc)
    elements.append(pie_drawing)

    # Table
    table_data = [["Utilisateurs", str(users_collection.count_documents({}))],
                  ["Offres", str(offers_collection.count_documents({}))],
                  ["Candidatures", str(applications_collection.count_documents({}))]]

    table = Table(table_data)
    table.setStyle(TableStyle([
        ('BACKGROUND', (0, 1), (-1, -1), colors.beige),
        ('GRID', (0, 0), (-1, -1), 1, colors.black)
    ]))
    elements.append(table)

    # User Information
    user_email = session.get('user_email', 'Utilisateur inconnu')
    user_info = Paragraph(f"Rapport généré par: {user_email}", styles['Normal'])
    elements.append(Spacer(1, 12))
    elements.append(user_info)

    doc.build(elements)

    buffer.seek(0)
    return send_file(buffer, as_attachment=True, download_name='statistiques.pdf', mimetype='application/pdf')

@app.route('/settings', methods=['GET', 'POST'])
def settings():
    if request.method == 'POST':
        user_email = session.get('user_email', '')
        new_full_name = request.form.get('full_name')
        new_password = request.form.get('new_password')

        update_data = {}
        if new_full_name:
            update_data['full_name'] = new_full_name
        if new_password:
            update_data['password'] = generate_password_hash(new_password)

        if update_data:
            users_collection.update_one({"email": user_email}, {"$set": update_data})

    user_email = session.get('user_email', '')
    user = users_collection.find_one({"email": user_email})

    return render_template_string(SETTINGS_TEMPLATE, user=user)

@app.route('/add_user', methods=['GET', 'POST'])
def add_user():
    if request.method == 'POST':
        email = request.form.get('email')
        full_name = request.form.get('full_name')
        user_type = request.form.get('user_type')
        password = request.form.get('password')

        hashed_password = generate_password_hash(password)

        new_user = {
            "email": email,
            "full_name": full_name,
            "user_type": user_type,
            "password": hashed_password
        }
        users_collection.insert_one(new_user)

        return redirect(url_for('dashboard', user_type='admin'))

    return render_template_string("""
<!DOCTYPE html>
<html lang="fr">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Ajouter un Utilisateur</title>
    <style>
        :root {
            --primary-color: #8052e6;
            --primary-dark: #6a40d0;
            --secondary-color: #2c2c6c;
            --accent-color: #ff6b6b;
            --success-color: #4ecdc4;
            --warning-color: #ffe66d;
            --background-color: #f8fafc;
            --card-background: #ffffff;
            --text-primary: #2d3748;
            --text-secondary: #718096;
            --border-color: #e2e8f0;
            --shadow-light: 0 4px 6px rgba(0, 0, 0, 0.07);
            --shadow-medium: 0 10px 25px rgba(0, 0, 0, 0.1);
            --shadow-heavy: 0 20px 40px rgba(0, 0, 0, 0.15);
            --border-radius: 12px;
            --transition: all 0.3s cubic-bezier(0.4, 0, 0.2, 1);
        }
        * {
            margin: 0;
            padding: 0;
            box-sizing: border-box;
        }
        body {
            font-family: 'Segoe UI', sans-serif;
            background: var(--background-color);
            color: var(--text-primary);
            line-height: 1.6;
        }
        .form-container {
            background: var(--card-background);
            padding: 2rem;
            border-radius: var(--border-radius);
            box-shadow: var(--shadow-light);
            border: 1px solid var(--border-color);
            transition: var(--transition);
            max-width: 600px;
            margin: 2rem auto;
        }
        .form-container h2 {
            color: var(--primary-color);
            margin-bottom: 1.5rem;
            text-align: center;
        }
        .form-group {
            margin-bottom: 1.5rem;
        }
        .form-group label {
            display: block;
            margin-bottom: 0.5rem;
            font-weight: 600;
            color: var(--text-primary);
        }
        .form-group input,
        .form-group select {
            width: 100%;
            padding: 1rem;
            border: 2px solid var(--border-color);
            border-radius: var(--border-radius);
            font-size: 1rem;
            transition: var(--transition);
        }
        .form-group input:focus,
        .form-group select:focus {
            outline: none;
            border-color: var(--primary-color);
            box-shadow: 0 0 0 3px rgba(128, 82, 230, 0.1);
        }
        .submit-button {
            background: linear-gradient(135deg, var(--primary-color), var(--primary-dark));
            color: white;
            border: none;
            padding: 1rem;
            border-radius: var(--border-radius);
            font-size: 1rem;
            font-weight: 600;
            cursor: pointer;
            transition: var(--transition);
            display: block;
            width: 100%;
            margin-top: 1rem;
        }
        .submit-button:hover {
            transform: translateY(-2px);
            box-shadow: var(--shadow-medium);
        }
    </style>
</head>
<body>
    <div class="form-container">
        <h2>Ajouter un Utilisateur</h2>
        <form method="post">
            <div class="form-group">
                <label for="email">E-mail</label>
                <input type="email" id="email" name="email" placeholder="Entrez votre e-mail" required>
            </div>
            <div class="form-group">
                <label for="full_name">Nom Complet</label>
                <input type="text" id="full_name" name="full_name" placeholder="Entrez votre nom complet" required>
            </div>
            <div class="form-group">
                <label for="user_type">Type d'utilisateur</label>
                <select id="user_type" name="user_type" required>
                    <option value="" disabled selected>Sélectionnez le type d'utilisateur</option>
                    <option value="freelance">Freelance</option>
                    <option value="recruiter">Recruteur</option>
                    <option value="admin">Administrateur</option>
                </select>
            </div>
            <div class="form-group">
                <label for="password">Mot de passe</label>
                <input type="password" id="password" name="password" placeholder="Entrez votre mot de passe" required>
            </div>
            <button type="submit" class="submit-button">Ajouter</button>
        </form>
    </div>
</body>
</html>
""")

@app.route('/edit_user/<user_id>', methods=['GET', 'POST'])
def edit_user(user_id):
    user = users_collection.find_one({"_id": ObjectId(user_id)})
    if not user:
        return "Utilisateur introuvable", 404

    if request.method == 'POST':
        full_name = request.form.get('full_name')
        user_type = request.form.get('user_type')
        password = request.form.get('password')

        update_data = {}
        if full_name:
            update_data['full_name'] = full_name
        if user_type:
            update_data['user_type'] = user_type
        if password:
            update_data['password'] = generate_password_hash(password)

        users_collection.update_one({"_id": ObjectId(user_id)}, {"$set": update_data})

        return redirect(url_for('dashboard', user_type='admin'))

    return render_template_string(f"""
<!DOCTYPE html>
<html lang="fr">
<head>
    <meta charset="UTF-8">
    <title>Modifier l'Utilisateur</title>
    <style>
        body {{
            background: #f8fafc;
            font-family: 'Segoe UI', sans-serif;
        }}
        .form-container {{
            background: #ffffff;
            max-width: 600px;
            margin: 3rem auto;
            padding: 2rem;
            border-radius: 12px;
            box-shadow: 0 10px 25px rgba(0,0,0,0.08);
        }}
        h2 {{
            color: #8052e6;
            margin-bottom: 1.5rem;
            text-align: center;
        }}
        .form-group {{
            margin-bottom: 1.2rem;
        }}
        label {{
            font-weight: bold;
            display: block;
            margin-bottom: 0.5rem;
            color: #333;
        }}
        input, select {{
            width: 100%;
            padding: 0.9rem;
            border: 2px solid #e2e8f0;
            border-radius: 8px;
            font-size: 1rem;
        }}
        input:focus, select:focus {{
            outline: none;
            border-color: #8052e6;
            box-shadow: 0 0 0 3px rgba(128, 82, 230, 0.1);
        }}
        .submit-button {{
            background: linear-gradient(135deg, #8052e6, #6a40d0);
            color: white;
            border: none;
            padding: 1rem;
            width: 100%;
            border-radius: 8px;
            font-weight: bold;
            font-size: 1rem;
            cursor: pointer;
            transition: all 0.3s ease;
        }}
        .submit-button:hover {{
            transform: translateY(-2px);
            box-shadow: 0 8px 25px rgba(0, 0, 0, 0.15);
        }}
    </style>
</head>
<body>
    <div class="form-container">
        <h2>✏️ Modifier l'Utilisateur</h2>
        <form method="post">
            <div class="form-group">
                <label for="full_name">Nom Complet :</label>
                <input type="text" id="full_name" name="full_name" value="{user['full_name']}" required>
            </div>
            <div class="form-group">
                <label for="user_type">Type d'utilisateur :</label>
                <select id="user_type" name="user_type">
                    <option value="freelance" {'selected' if user['user_type'] == 'freelance' else ''}>Freelance</option>
                    <option value="recruiter" {'selected' if user['user_type'] == 'recruiter' else ''}>Recruteur</option>
                    <option value="admin" {'selected' if user['user_type'] == 'admin' else ''}>Administrateur</option>
                </select>
            </div>
            <div class="form-group">
                <label for="password">Nouveau Mot de Passe :</label>
                <input type="password" id="password" name="password" placeholder="Laisser vide pour ne pas changer">
            </div>
            <button type="submit" class="submit-button">✅ Enregistrer les modifications</button>
        </form>
    </div>
</body>
</html>
""")

from bson.objectid import ObjectId

@app.route('/delete_user/<user_id>', methods=['DELETE'])
def delete_user(user_id):
    print("🧨 Suppression utilisateur ID :", user_id)
    try:
        result = users_collection.delete_one({"_id": ObjectId(user_id)})
        if result.deleted_count == 1:
            return jsonify({"success": True})
        else:
            return jsonify({"success": False, "message": "Utilisateur introuvable."})
    except Exception as e:
        print("❌ Erreur :", e)
        return jsonify({"success": False, "message": "Erreur lors de la suppression."})



@app.route('/ask_ai_with_matching', methods=['POST'])
def ask_ai_with_matching():
    user_id = session.get('user_id')
    user = users_collection.find_one({'_id': ObjectId(user_id)})

    if not user:
        return jsonify({'answer': "Utilisateur non authentifié."})

    question = request.json.get("question", "")

    if user['user_type'] == 'freelance':
        return handle_student_matching(user_id, question)

    elif user['user_type'] == 'recruiter':
        return handle_recruiter_matching(user_id, question)

    return jsonify({'answer': "Profil non reconnu."})
from datetime import datetime, timezone

def handle_student_matching(user_id, question):
    user = users_collection.find_one({"_id": ObjectId(user_id)})
    email = user.get("email")

    # 1. Récupérer le contenu du CV
    cv_doc = cvs_collection.find_one({"email": email}, sort=[("uploaded_at", -1)])
    if not cv_doc or "content" not in cv_doc:
        return jsonify({"error": "Aucun contenu de CV trouvé."})
    cv_text = cv_doc["content"]

    # 2. Récupérer les projets de l'utilisateur
    projects = projects_collection.find({"user_id": user_id})
    projects_text = "\n".join([
        f"- {proj['title']} : {proj.get('description', '')}, Techs: {proj.get('technologies', '')}"
        for proj in projects
    ]) or "Aucun projet enregistré."

    # 3. Sélectionner les offres
    offers = list(offers_collection.find().sort("created_at", -1).limit(3))
    if not offers:
        return jsonify({"error": "Aucune offre en base de données."})

    offer_ids = [str(offer['_id']) for offer in offers]
    offres_formatees = ""
    for i, offer in enumerate(offers):
        offres_formatees += f"\n🔹 Offre {i+1}:\nTitre: {offer.get('title', 'Sans titre')}\nDescription: {offer.get('description', '...')}"

    # 4. Prompt IA
    prompt = f"""
Tu es un expert en recrutement IA tu doit donner le resultat en bref pour ne pas souler l'utilisateur de lire. Voici le profil à analyser :

🎓 CV : {cv_text}
📁 Projets : {projects_text}
📄 Offres à évaluer : {offres_formatees}

Ta mission :
- Analyse chaque offre avec ce CV
- Fais une évaluation pondérée :
  • Compétences 25% 
  • Expérience 25%
  • Formation 25%
  • Projets 15%
  • Certifications 5%
  • Softskills 5%
- Pour chaque offre, indique :
  ✅ Matching (oui/non)
  💯 Score /100
  📝 Recommandation (1 ligne)
  🧠 Résumé clair en 5 lignes max
"""

    headers = {
        "Authorization": "Bearer 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",
        "Content-Type": "application/json"
    }

    payload = {
        "model": "meta-llama/llama-4-scout-17b-16e-instruct",
        "messages": [{"role": "user", "content": prompt}]
    }

    try:
        response = requests.post(
            "https://api.groq.com/openai/v1/chat/completions",
            headers=headers,
            json=payload
        )
        result = response.json()
        answer = result["choices"][0]["message"]["content"]

        # ✅ Enregistrement dans MongoDB avec datetime valide
        matching_results_collection.insert_one({
            "user_id": user_id,
            "question": question,
            "response": answer,
            "offer_ids": offer_ids,
            "timestamp": datetime.now(timezone.utc)
        })

        return jsonify({"answer": answer})
    except Exception as e:
        return jsonify({"error": str(e)})



@app.route('/generate_reports')
def generate_reports():
    import datetime
    import pandas as pd
    import matplotlib.pyplot as plt

    # 1. Statistiques globales
    stats = {
        "Utilisateurs": users_collection.count_documents({}),
        "Offres": offers_collection.count_documents({}),
        "Offres scrapées": scraped_offers_collection.count_documents({}),
        "Messages": messages_collection.count_documents({}),
        "CVs": cvs_collection.count_documents({}),
        "Candidatures": applications_collection.count_documents({}),
        "Projets": projects_collection.count_documents({}),
        "Stats": stats_collection.count_documents({}),
        "Matching IA": matching_results_collection.count_documents({}),
        "Profils Étudiants": student_profiles_collection.count_documents({}),
        "Notifications": notifications_collection.count_documents({})
    }

    # 2. Génération CSV et PDF dans /static
    df = pd.DataFrame(list(stats.items()), columns=["Paramètre", "Quantité"])
    df.to_csv("static/rapport_stats.csv", index=False)

    plt.figure(figsize=(8, 8))
    plt.pie(df["Quantité"], labels=df["Paramètre"], autopct="%1.1f%%")
    plt.axis("equal")
    plt.title("Répartition des entités Web4Jobs")
    plt.tight_layout()
    plt.savefig("static/rapport_stats.pdf")

    labels = list(stats.keys())
    data = list(stats.values())

    # 3. Historique matching IA
    results = matching_results_collection.find().sort("timestamp", -1)
    html_blocks = []

    for res in results:
        user_id = res.get("user_id")
        timestamp = res.get("timestamp")
        response = res.get("response", "").replace('\n', '<br>')

        try:
            user = users_collection.find_one({"_id": ObjectId(user_id)}) if user_id else None
            user_name = user.get("full_name", "Utilisateur inconnu") if user else "Utilisateur inconnu"
        except:
            user_name = "Utilisateur inconnu"

        date_str = timestamp.strftime("%d/%m/%Y %H:%M") if isinstance(timestamp, datetime.datetime) else "Date inconnue"

        html_blocks.append(f"""
        <div style="border:1px solid #ddd; padding:1rem; border-radius:10px; margin-bottom:1rem; background:#f9f9f9;">
            <h3>🧑 {user_name}</h3>
            <p><strong>📅 Date :</strong> {date_str}</p>
            <p><strong>🧠 Résultat IA :</strong><br>{response}</p>
        </div>
        """)

    # 4. HTML final avec Chart.js + téléchargements
    full_html = f"""
    <!DOCTYPE html>
    <html lang="fr">
    <head>
        <meta charset="UTF-8">
        <title>Rapports Web4Jobs</title>
        <script src="https://cdn.jsdelivr.net/npm/chart.js"></script>
        <style>
            body {{
                font-family: 'Segoe UI', sans-serif;
                background-color: #f0f2f5;
                padding: 2rem;
                color: #333;
            }}
            h1 {{
                text-align: center;
                color: #8052e6;
                margin-bottom: 2rem;
            }}
            .chart-container {{
                max-width: 600px;
                margin: 0 auto 2rem;
            }}
            .download-links {{
                text-align: center;
                margin-bottom: 2rem;
            }}
            .download-links a {{
                display: inline-block;
                margin: 0.5rem;
                background: #8052e6;
                color: white;
                padding: 0.5rem 1rem;
                border-radius: 8px;
                text-decoration: none;
            }}
            .download-links a:hover {{
                background: #6a40d0;
            }}
        </style>
    </head>
    <body>
        <h1>📊 Rapports de l'application Web4Jobs</h1>

        <div class="chart-container">
            <canvas id="statsChart"></canvas>
        </div>

        <div class="download-links">
            <a href="/static/rapport_stats.csv" download>📥 Télécharger CSV</a>
            <a href="/static/rapport_stats.pdf" download>📥 Télécharger PDF</a>
        </div>

        <h2>🧠 Historique des Matchings IA</h2>
        {''.join(html_blocks)}

        <script>
            const ctx = document.getElementById('statsChart').getContext('2d');
            new Chart(ctx, {{
                type: 'pie',
                data: {{
                    labels: {labels},
                    datasets: [{{
                        label: "Répartition des entités",
                        data: {data},
                        borderWidth: 1
                    }}]
                }},
                options: {{
                    responsive: true,
                    plugins: {{
                        legend: {{
                            position: 'bottom',
                            labels: {{
                                color: '#333',
                                font: {{
                                    size: 14
                                }}
                            }}
                        }}
                    }}
                }}
            }});
        </script>
    </body>
    </html>
    """
    return full_html




@app.route('/api/stats')
def get_stats():
    from_date = lambda y, m: datetime(y, m, 1)

    year = int(request.args.get('year', datetime.now().year))
    month = int(request.args.get('month', datetime.now().month))

    current_start = from_date(year, month)
    next_month = month + 1 if month < 12 else 1
    next_year = year if month < 12 else year + 1
    current_end = from_date(next_year, next_month)

    prev_month = 12 if month == 1 else month - 1
    prev_year = year - 1 if month == 1 else year
    prev_start = from_date(prev_year, prev_month)
    prev_end = from_date(year, month)

    def count(col, extra={}):
        return {
            "current": db[col].count_documents({**{"createdAt": {"$gte": current_start, "$lt": current_end}}, **extra}),
            "previous": db[col].count_documents({**{"createdAt": {"$gte": prev_start, "$lt": prev_end}}, **extra})
        }

    return jsonify({
        "users": count("users"),
        "offers": count("offers"),
        "applications": count("applications"),
        "valid": count("applications", {"status": "validée"}),
        "refused": count("applications", {"status": "refusée"}),
        "matching_ia": count("match_stats")
    })

@app.route('/api/stats_trend')
def stats_trend():
    from datetime import timedelta
    today = datetime.now()
    months_data = []

    for i in range(6)[::-1]:  # les 6 derniers mois
        ref = today.replace(day=1) - timedelta(days=30 * i)
        y, m = ref.year, ref.month
        start = datetime(y, m, 1)
        end_month = 1 if m == 12 else m + 1
        end_year = y + 1 if m == 12 else y
        end = datetime(end_year, end_month, 1)

        def count(col, extra={}):
            return db[col].count_documents({**{"createdAt": {"$gte": start, "$lt": end}}, **extra})

        months_data.append({
            "label": f"{start.strftime('%b')} {y}",
            "users": count("users"),
            "offers": count("offers"),
            "applications": count("applications"),
            "matching_ia": count("match_stats")
        })

    return jsonify(months_data)

@app.route('/recruteur/stats')
def recruiter_stats_page():
    recruiter_id = session.get('user_id')

    nb_offres = offers_collection.count_documents({'recruiter_id': recruiter_id})
    nb_candidatures = applications_collection.count_documents({'recruiter_id': recruiter_id})
    nb_projets = projects_collection.count_documents({'recruiter_id': recruiter_id, 'status': 'en cours'})

    top_skills = [
        ("Python", 5),
        ("SQL", 4),
        ("ML", 3),
        ("Flask", 2),
        ("FastAPI", 1)
    ]

    return render_template("recruiter_stats.html",
                           nb_offres=nb_offres,
                           nb_candidatures=nb_candidatures,
                           nb_projets=nb_projets,
                           top_skills=top_skills)


@app.route('/recruteur/offers')
def recruiter_offers():
    recruiter_id = session.get('user_id')
    offres = list(offers_collection.find({'recruiter_id': recruiter_id}))
    for o in offres:
        o['_id'] = str(o['_id'])
    return jsonify(offres)

@app.route('/recruteur/applications')
def recruiter_applications():
    recruiter_id = session.get('user_id')
    pipeline = [
        {"$match": {"recruiter_id": recruiter_id}},
        {"$lookup": {
            "from": "users",
            "localField": "student_id",
            "foreignField": "_id",
            "as": "student_info"
        }},
        {"$unwind": "$student_info"},
        {"$project": {
            "student_name": "$student_info.full_name",
            "student_email": "$student_info.email"
        }}
    ]
    candidatures = list(applications_collection.aggregate(pipeline))
    return jsonify(candidatures)

@app.route('/recherche-talents', methods=['GET'])
def recherche_talents():
    query = request.args.get('q', '')
    if query:
        regex_query = {'$regex': query, '$options': 'i'}
        profils = list(db.student_profiles.find({
            '$or': [
                {'nom': regex_query},
                {'email': regex_query},
                {'competences': regex_query}
            ]
        }))
    else:
        profils = []

    return render_template_string("""
<!DOCTYPE html>
<html lang="fr">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Recherche de Talents</title>
    <style>
        body {
            font-family: Arial, sans-serif;
            background-color: #f4f4f9;
            margin: 0;
            padding: 20px;
        }
        .search-container {
            max-width: 800px;
            margin: 0 auto;
            background: #fff;
            padding: 20px;
            border-radius: 8px;
            box-shadow: 0 0 10px rgba(0, 0, 0, 0.1);
        }
        .search-form {
            display: flex;
            margin-bottom: 20px;
        }
        .search-form input[type="text"] {
            flex: 1;
            padding: 10px;
            border: 1px solid #ddd;
            border-radius: 4px;
        }
        .search-form button {
            padding: 10px 20px;
            background-color: #6200ee;
            color: white;
            border: none;
            border-radius: 4px;
            margin-left: 10px;
            cursor: pointer;
        }
        .search-form button:hover {
            background-color: #7b1fa2;
        }
        .results {
            margin-top: 20px;
        }
        .profile {
            background: #f9f9f9;
            padding: 15px;
            margin-bottom: 10px;
            border-radius: 4px;
            border-left: 4px solid #6200ee;
        }
        .profile h3 {
            margin-top: 0;
            color: #6200ee;
        }
    </style>
</head>
<body>
    <div class="search-container">
        <h2>Recherche de Talents</h2>
        <form class="search-form" action="/recherche-talents" method="get">
            <input type="text" name="q" placeholder="Rechercher des talents..." value="{{ query }}">
            <button type="submit">Rechercher</button>
        </form>

        <div class="results">
            {% if profils %}
                {% for profil in profils %}
                    <div class="profile">
                        <h3>{{ profil.nom }}</h3>
                        <p><strong>Email:</strong> {{ profil.email }}</p>
                        <p><strong>Compétences:</strong> {{ ', '.join(profil.get('competences', [])) }}</p>
                    </div>
                {% endfor %}
            {% elif query %}
                <p>Aucun résultat trouvé pour "{{ query }}".</p>
            {% else %}
                <p>Veuillez entrer un terme de recherche.</p>
            {% endif %}
        </div>
    </div>
</body>
</html>
""", profils=profils, query=query)

@app.route('/recruteur/statistiques')
def statistiques_recruteur():
    recruiter_email = session.get('user_email')

    nb_offres = offers_collection.count_documents({'created_by': recruiter_email})
    nb_candidatures = applications_collection.count_documents({'recruiter_email': recruiter_email})
    nb_projets = projects_collection.count_documents({'email': recruiter_email})

    skills_stats = list(db.student_profiles.aggregate([
        {"$unwind": "$competences"},
        {"$group": {"_id": "$competences", "count": {"$sum": 1}}},
        {"$sort": {"count": -1}},
        {"$limit": 5}
    ]))

    top_skills = [(s["_id"], s["count"]) for s in skills_stats]

    return render_template_string(RECRUITER_STATS_TEMPLATE,
                           nb_offres=nb_offres,
                           nb_candidatures=nb_candidatures,
                           nb_projets=nb_projets,
                           top_skills=top_skills)

@app.route('/recruteur/settings', methods=['GET', 'POST'])
def settings_recruteur():
    if 'user_email' not in session:
        return redirect('/login')

    recruiter_email = session['user_email']
    recruteur = users_collection.find_one({'email': recruiter_email})

    if request.method == 'POST':
        nom = request.form.get('nom')
        description = request.form.get('description')
        password = request.form.get('password')

        update_data = {
            'nom': nom,
            'description': description,
        }

        if password:
            update_data['password'] = generate_password_hash(password)

        users_collection.update_one({'email': recruiter_email}, {'$set': update_data})
        return redirect('/recruteur/settings')

    return render_template_string(RECRUITER_SETTINGS_TEMPLATE, recruteur=recruteur)

@app.route('/recruiter/recent_candidates')
def recruiter_recent_candidates():
    from bson import ObjectId
    from datetime import datetime
    recruiter_id = session.get("user_id")

    if not recruiter_id:
        return jsonify({"error": "Non connecté"}), 401

    # Offres de ce recruteur
    recruiter_offers = offers_collection.find({"recruiter_id": recruiter_id})
    offer_ids = [str(offer["_id"]) for offer in recruiter_offers]

    # Candidatures sur ces offres
    applications = applications_collection.find({
        "offer_id": {"$in": offer_ids}
    }).sort("created_at", -1).limit(5)

    results = []
    for app in applications:
        user = users_collection.find_one({"_id": ObjectId(app["user_id"])})
        offer = offers_collection.find_one({"_id": ObjectId(app["offer_id"])})

        if user and offer:
            results.append({
                "candidate_name": user.get("name", "Inconnu"),
                "candidate_email": user.get("email", ""),
                "offer_title": offer.get("title", "Sans titre"),
                "status": app.get("status", "À examiner"),
                "applied_at": app.get("created_at", str(datetime.now()))
            })

    return jsonify(results)


@app.route('/api/recruiter/candidates')
def fetch_recent_candidates():
    candidates = list(student_profiles_collection.find().sort('_id', -1).limit(5))
    for c in candidates:
        c['_id'] = str(c['_id'])
    return jsonify(candidates)


@app.route('/api/recruiter/projects')
def get_active_projects():
    projects = list(projects_collection.find().sort('created_at', -1).limit(5))
    for p in projects:
        p['_id'] = str(p['_id'])
    return jsonify(projects)


from flask import render_template_string

@app.route('/recruiter/scrape_remoteok')
def recruiter_scrape_remoteok():
    query = request.args.get("q", "python")
    location = request.args.get("l", "remote")
    offers = scrape_remoteok(query)
    save_scraped_offers(offers)
    return render_template_string(recruiter_scraped_offers, offers=offers)





def scrape_remoteok(query="python"):
    import requests
    import time

    url = f"https://remoteok.com/remote-{query}-jobs"
    headers = {
        "User-Agent": "Mozilla/5.0"
    }
    response = requests.get(url, headers=headers)

    if response.status_code != 200:
        print("[RemoteOK] Échec du scraping")
        return []

    from bs4 import BeautifulSoup
    soup = BeautifulSoup(response.text, "html.parser")
    jobs = []

    for row in soup.find_all("tr", class_="job"):
        title = row.find("h2")
        company = row.find("h3")
        tags = [tag.text for tag in row.find_all("div", class_="tag")]

        jobs.append({
            "source": "RemoteOK",
            "title": title.text.strip() if title else "Sans titre",
            "company": company.text.strip() if company else "Inconnu",
            "description": f"Tags: {', '.join(tags)}",
            "tags": tags,
            "location": "Remote"
        })

    print(f"[RemoteOK] {len(jobs)} offres récupérées.")
    return jobs

def save_scraped_offers(offers):
    for offer in offers:
        if not scraped_offers_collection.find_one({"title": offer["title"], "company": offer["company"]}):
            scraped_offers_collection.insert_one(offer)

@app.route('/recruiter/publish_scraped_offer', methods=["POST"])
def publish_scraped_offer():
    title = request.form.get("title")
    description = request.form.get("description")
    tags = request.form.get("tags", "").split(",")

    offers_collection.insert_one({
        "title": title,
        "description": description,
        "tags": tags,
        "source": "scraped",
        "status": "active",
        "created_at": datetime.utcnow()
    })

    return redirect('/recruteur/dashboard')
#............................................................................................
@app.route("/generate_pdf_report/<user_id>")
def generate_pdf_report(user_id):
    result = matching_results_collection.find_one({"user_id": user_id})
    if not result:
        return "Aucun résultat trouvé pour ce profil", 404

    scores = result.get("scores", {})
    offers = offers_collection.find({"_id": {"$in": [ObjectId(oid) for oid in scores.keys()]}})
    offer_titles = {str(offer["_id"]): offer["title"] for offer in offers}

    # Création du diagramme circulaire
    labels = [offer_titles.get(oid, oid[:6]) for oid in scores]
    values = list(scores.values())

    fig, ax = plt.subplots()
    ax.pie(values, labels=labels, autopct='%1.1f%%', startangle=90)
    ax.axis('equal')
    
    img_path = tempfile.mktemp(suffix=".png")
    plt.savefig(img_path)
    plt.close()

    # Création du PDF
    buffer = tempfile.NamedTemporaryFile(delete=False, suffix=".pdf")
    doc = SimpleDocTemplate(buffer.name, pagesize=A4)
    styles = getSampleStyleSheet()
    elements = []

    elements.append(Paragraph(f"Rapport de Matching - Utilisateur : {user_id}", styles['Title']))
    elements.append(Paragraph(f"Date : {datetime.now().strftime('%d/%m/%Y %H:%M')}", styles['Normal']))
    elements.append(Spacer(1, 12))

    data = [["Offre", "Score"]]
    for oid, score in scores.items():
        title = offer_titles.get(oid, f"Offre {oid[:6]}")
        data.append([title, f"{score}/100"])

    table = Table(data)
    table.setStyle(TableStyle([
        ('BACKGROUND', (0,0), (-1,0), colors.lightgrey),
        ('TEXTCOLOR',(0,0),(-1,0),colors.black),
        ('ALIGN',(0,0),(-1,-1),'CENTER'),
        ('GRID', (0,0), (-1,-1), 1, colors.black),
    ]))
    elements.append(table)
    elements.append(Spacer(1, 12))
    elements.append(Image(img_path, width=400, height=300))

    doc.build(elements)
    os.remove(img_path)

    return send_file(buffer.name, as_attachment=True, download_name="rapport_matching.pdf")

from flask import render_template_string, session, redirect
from bson import ObjectId

@app.route("/student/dashboard")
def student_dashboard():
    if 'user_id' not in session:
        return redirect("/login")

    user_id = ObjectId(session["user_id"])

    # 🔢 Nombre de candidatures
    nb_candidatures = applications_collection.count_documents({"user_id": user_id})

    # 🔢 Nombre total de profils étudiants
    nb_etudiants = student_profiles_collection.count_documents({})

    # 📄 Dernier CV
    latest_cv = cvs_collection.find_one({"user_id": user_id}, sort=[("uploaded_at", -1)])
    latest_cv_filename = latest_cv['filename'] if latest_cv else None

    return render_template_string(STUDENT_DASHBOARD_TEMPLATE,
                                  nb_candidatures=nb_candidatures,
                                  nb_etudiants=nb_etudiants,
                                  latest_cv=latest_cv_filename)


@app.route("/test_etudiants")
def test_etudiants():
    nb = student_profiles_collection.count_documents({})
    return f"<h2>Il y a actuellement {nb} profil(s) étudiant(s) en base.</h2>"

# ➕ Générer fichiers statiques à télécharger
import pandas as pd
import matplotlib.pyplot as plt

@app.route('/recruiter/stats_report')
def recruiter_stats_report():
    import pandas as pd
    import matplotlib.pyplot as plt

    stats = {
        "Utilisateurs": users_collection.count_documents({"user_type": "recruiter"}),
        "Offres publiées": offers_collection.count_documents({}),
        "Projets déposés": projects_collection.count_documents({}),
        "Matchings IA": matching_results_collection.count_documents({}),
        "Profils étudiants": student_profiles_collection.count_documents({})
    }

    # Générer fichier CSV
    df = pd.DataFrame(list(stats.items()), columns=["Paramètre", "Quantité"])
    df.to_csv("static/recruteur_stats.csv", index=False)

    # Générer fichier PDF circulaire
    plt.figure(figsize=(8, 8))
    plt.pie(df["Quantité"], labels=df["Paramètre"], autopct="%1.1f%%", startangle=140)
    plt.axis("equal")
    plt.title("📊 Répartition des statistiques Recruteur")
    plt.tight_layout()
    plt.savefig("static/recruteur_stats.pdf")
    plt.close()

    # Préparation des données pour Chart.js
    labels = list(stats.keys())
    values = list(stats.values())

    return render_template_string("""
    <!DOCTYPE html>
    <html lang="fr">
    <head>
        <meta charset="UTF-8">
        <title>📊 Statistiques Recruteur</title>
        <script src="https://cdn.jsdelivr.net/npm/chart.js"></script>
        <style>
            body { font-family: sans-serif; background: #f4f6fa; padding: 2rem; }
            h1 { color: #764ba2; text-align: center; margin-bottom: 2rem; }
            .container { max-width: 700px; margin: auto; text-align: center; }
            .downloads a {
                display: inline-block;
                margin: 1rem;
                padding: 0.7rem 1.2rem;
                background: #764ba2;
                color: white;
                text-decoration: none;
                border-radius: 8px;
            }
            canvas { margin-top: 2rem; }
        </style>
    </head>
    <body>
        <h1>📊 Statistiques du Recruteur</h1>
        <div class="container">
            <div class="downloads">
                <a href="/static/recruteur_stats.csv" download>Télécharger CSV</a>
                <a href="/static/recruteur_stats.pdf" download>Télécharger PDF</a>
            </div>
            <canvas id="recruiterChart"></canvas>
        </div>
        <script>
            const ctx = document.getElementById('recruiterChart').getContext('2d');
            new Chart(ctx, {
                type: 'pie',
                data: {
                    labels: {{ labels | safe }},
                    datasets: [{
                        label: 'Statistiques',
                        data: {{ values | safe }},
                        backgroundColor: ['#8052e6', '#6a40d0', '#4ecdc4', '#ff6b6b', '#ffe66d']
                    }]
                },
                options: {
                    responsive: true,
                    plugins: {
                        legend: {
                            position: 'bottom'
                        }
                    }
                }
            });
        </script>
    </body>
    </html>
    """, labels=labels, values=values)
#...........................................................................notification.....
@app.route('/api/get_notifications')
def get_notifications():
    role = session.get('user_type')
    if not role:
        return jsonify([])

    notifications = list(notifications_collection.find({"target_role": role}).sort("timestamp", -1))
    return jsonify([
        {
            "message": n["message"],
            "timestamp": n["timestamp"].strftime('%d/%m/%Y %H:%M')
        } for n in notifications
    ])
#..............................................................................postuler......
@app.route('/postuler', methods=['GET'])
def postuler():
    if 'user_email' not in session:
        return redirect('/login')

    email = session['user_email']
    student_profile = student_profiles_collection.find_one({"email": email})
    matching_result = matching_results_collection.find_one({"email": email})

    if not student_profile or not matching_result:
        return "Profil ou résultats de matching introuvables.", 404

    full_name = student_profile.get('full_name', 'Étudiant inconnu')
    cv_link = student_profile.get('cv_url', '#')
    result_summary = matching_result.get('match_result', 'Pas de résultats.')

    message = f"📨 Candidature reçue de {full_name}.\n🧠 Résumé matching : {result_summary[:100]}..."

    add_notification(
        role='recruiter',
        message=message,
        student_email=email,
        cv_url=cv_link
    )

    return redirect('/dashboard_student')

#...................................................................................../dashboard_student
@app.route('/dashboard_student')
def dashboard_student():
    if 'user_email' not in session:
        return redirect('/login')

    email = session['user_email']
    user = users_collection.find_one({'email': email})
    if not user:
        return redirect('/login')

    # récupérer les stats de candidatures
    total_apps = applications_collection.count_documents({'email': email})

    return render_template_string(STUDENT_DASHBOARD_TEMPLATE, total_applications=total_apps)


#...................................................bouton cadidant 
@app.route('/candidats_projets')
def voir_candidats_projets():
    if 'user_type' not in session or session['user_type'] != 'recruiter':
        return redirect('/login')

    try:
        candidats = list(student_profiles_collection.find())
        projets = list(projects_collection.find())
        matchings = list(matching_results_collection.find())
        cvs = list(cvs_collection.find())
        notifications = list(notifications_collection.find({"target_role": "recruiter"}))

        data_by_email = {}

        # 1. Initialiser les profils candidats
        for c in candidats:
            email = c.get('email')
            if not email:
                continue
            data_by_email[email] = {
                "profile": c,
                "projects": [],
                "matching": None,
                "cv": None,
                "status": "En cours"
            }

        # 2. Ajouter les projets
        for p in projets:
            email = p.get('email')
            if email in data_by_email:
                data_by_email[email]["projects"].append(p)

        # 3. Ajouter les résultats de matching
        for m in matchings:
            email = m.get('email')
            if email in data_by_email and 'match_result' in m:
                data_by_email[email]["matching"] = {
                    "match_result": m["match_result"]
                }

        # 4. Ajouter les liens de CV
        for cv in cvs:
            email = cv.get('email')
            if email in data_by_email and 'url' in cv:
                data_by_email[email]["cv"] = {
                    "url": cv["url"]
                }

        # 5. Ajouter les statuts de candidature
        for notif in notifications:
            email = notif.get('student_email')
            if email in data_by_email and 'status' in notif:
                data_by_email[email]["status"] = notif['status']

        return render_template_string(CANDIDATS_PROJETS_TEMPLATE, data=data_by_email)

    except Exception as e:
        return f"<h2>Erreur détectée : {str(e)}</h2>", 500


#......................................................../update_candidate_status
@app.route('/update_candidate_status', methods=['POST'])
def update_candidate_status():
    data = request.json
    email = data.get('email')
    new_status = data.get('status')

    if not email or not new_status:
        return jsonify({'success': False, 'message': 'Données manquantes'})

    notifications_collection.update_many(
        {'student_email': email},
        {'$set': {'status': new_status}}
    )

    return jsonify({'success': True, 'message': f"Statut mis à jour en {new_status}"})



#................................................................................
@app.route('/btnMatching')
def btnMatching():
    # Appel indirect à la logique de /matching_results en redirection
    return redirect(url_for("matching_results"))

if __name__ == '__main__':
    app.run(host="0.0.0.0", port=5000, debug=False)


```
